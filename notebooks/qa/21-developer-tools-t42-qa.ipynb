```python
# NORTHSTAR -- Tower 42 Developer Tools QA for Solace Browser
# DNA: devtools(suite) = issues(triage) x pr(review) x alert(notify) x priority(rank) x assign(auto) x report(generate) x integrate(ci)
# Probes issue scanning, PR review, notifications, priority ranking, CI integration
import os, sys, re, json, hashlib
from pathlib import Path
from datetime import datetime

NORTHSTAR = "developer-tools-t42-qa"
NOTEBOOK_PATH = "notebooks/qa/21-developer-tools-t42-qa.ipynb"
TOWER = 42
TOWER_NAME = "Tower of Developer Tools"
MIN_SCORE = 70

BROWSER_ROOT = Path("/home/phuc/projects/solace-browser")
SRC = BROWSER_ROOT / "src"
WEB = BROWSER_ROOT / "web"
TESTS = BROWSER_ROOT / "tests"
DATA = BROWSER_ROOT / "data" / "default"

results = []

def record(probe_id, name, passed, detail=""):
    status = "PASS" if passed else "FAIL"
    results.append({"id": probe_id, "name": name, "status": status, "detail": detail})
    print(f"{status}: {name}" + (f" -- {detail}" if detail else ""))

def read_file(path):
    p = Path(path)
    return p.read_text(encoding='utf-8', errors='replace') if p.exists() else ""

print(f"Tower {TOWER}: {TOWER_NAME}")
assert BROWSER_ROOT.exists()
```

```python
# F1 -- Alex Petrov: Issue Scanning (Lead Developer)
# Q: Can I see all GitHub issues across repos in unified view?
# Developer tools are implemented via recipes and companion apps

# Probe: recipe system can handle GitHub-style data extraction
recipe_files = list(DATA.rglob("*.json")) if DATA.exists() else []
github_recipes = [f for f in recipe_files if 'github' in f.name.lower() or 'git' in f.name.lower()]
record("F1-P1", "GitHub-related recipes exist", len(github_recipes) > 0,
       f"{len(github_recipes)} GitHub recipes: {[f.name for f in github_recipes[:5]]}")

# Probe: companion bridge handles app data flow
companion_bridge = read_file(SRC / "companion" / "bridge.py")
has_bridge = bool(re.search(r'bridge|connect|data|send|receive', companion_bridge, re.IGNORECASE))
record("F1-P2", "Companion bridge for app data flow exists", has_bridge,
       f"{len(companion_bridge)} bytes")

# Probe: cross-app messaging exists
cross_msg = read_file(SRC / "cross_app" / "message.py")
has_messaging = bool(re.search(r'message|send|receive|dispatch|event', cross_msg, re.IGNORECASE))
record("F1-P3", "Cross-app messaging module exists", has_messaging,
       f"{len(cross_msg)} bytes")
```

```python
# F2 -- Priya Shankar: Priority Ranking (Product Manager)
# Q: Does the app rank issues by priority? Can I override?

# Probe: action engine can interact with web pages for data extraction
action_code = read_file(SRC / "action_engine.py")
has_extract = bool(re.search(r'extract|scrape|parse|data|content', action_code, re.IGNORECASE))
record("F2-P1", "Action engine can extract page data", has_extract,
       f"{len(action_code)} bytes")

# Probe: channels system for app data routing
channels_dir = SRC / "channels"
channel_files = list(channels_dir.glob("*.py")) if channels_dir.exists() else []
record("F2-P2", "Channels system exists for data routing",
       len(channel_files) >= 1, f"{len(channel_files)} channel files")

# Probe: channels test exists
channel_test = TESTS / "test_channels.py"
has_chan_test = channel_test.exists() and channel_test.stat().st_size > 100
record("F2-P3", "Channels test exists", has_chan_test)
```

```python
# F3 -- PR Review + F4 -- Alert/Notify
# Q: Can the browser navigate PRs? Can it generate review comments?

# Probe: browser layers for multi-page navigation
layers_code = read_file(SRC / "browser_layers.py")
has_layers = bool(re.search(r'layer|tab|window|navigate|page', layers_code, re.IGNORECASE))
record("F3-P1", "Browser layers for multi-page navigation exists", has_layers,
       f"{len(layers_code)} bytes")

# Probe: LLM clients exist for AI-powered review
claude_client = read_file(SRC / "llm" / "claude_code_client.py")
together_client = read_file(SRC / "llm" / "together_client.py")
has_llm = len(claude_client) > 100 or len(together_client) > 100
record("F3-P2", "LLM clients exist for AI-powered analysis", has_llm)

# Probe: alert queue for notifications
alert_code = read_file(SRC / "yinyang" / "alert_queue.py")
has_alerts = bool(re.search(r'alert|queue|notify|push|message', alert_code, re.IGNORECASE))
record("F4-P1", "Alert queue for developer notifications exists", has_alerts,
       f"{len(alert_code)} bytes")

# Probe: alert test exists
alert_test = TESTS / "test_alert_queue.py"
has_alert_test = alert_test.exists() and alert_test.stat().st_size > 100
record("F4-P2", "Alert queue test exists", has_alert_test)
```

```python
# F5 -- Reporting + CI Integration
# Q: Can the system generate reports from collected data?
# Q: Is there CI integration infrastructure?

# Probe: execution lifecycle tracks full task workflow
lifecycle = read_file(SRC / "execution_lifecycle.py")
has_lifecycle = bool(re.search(r'lifecycle|start|complete|fail|report|summary', lifecycle, re.IGNORECASE))
record("F5-P1", "Execution lifecycle tracks task workflow", has_lifecycle,
       f"{len(lifecycle)} bytes")

# Probe: machine API for headless CI usage
machine_api = read_file(SRC / "machine" / "api.py")
has_machine = bool(re.search(r'api|endpoint|route|headless|machine', machine_api, re.IGNORECASE))
record("F5-P2", "Machine API for headless/CI usage exists", has_machine,
       f"{len(machine_api)} bytes")

# Probe: machine test exists
machine_test = TESTS / "test_machine.py"
has_machine_test = machine_test.exists() and machine_test.stat().st_size > 100
record("F5-P3", "Machine/CI test exists", has_machine_test)

# Probe: machine dashboard page exists
dashboard_html = read_file(WEB / "machine-dashboard.html")
has_dashboard = len(dashboard_html) > 200
record("F5-P4", "Machine dashboard page exists", has_dashboard,
       f"{len(dashboard_html)} bytes")
```

```python
# F6 -- Assignment + F7 -- Integration
# Q: Can recipes be chained for complex dev workflows?

# Probe: recipe compiler supports chained steps
compiler_code = read_file(SRC / "recipes" / "recipe_compiler.py")
has_chain = bool(re.search(r'chain|sequence|step|pipeline|compose', compiler_code, re.IGNORECASE))
record("F6-P1", "Recipe compiler supports chained workflows", has_chain)

# Probe: inbox/outbox for async task handling
inbox_code = read_file(SRC / "inbox_outbox.py")
has_inbox = bool(re.search(r'inbox|outbox|queue|pending|process', inbox_code, re.IGNORECASE))
record("F6-P2", "Inbox/outbox system for async task handling exists", has_inbox,
       f"{len(inbox_code)} bytes")

# Probe: inbox/outbox test exists
inbox_test = TESTS / "test_inbox_outbox.py"
has_inbox_test = inbox_test.exists() and inbox_test.stat().st_size > 100
record("F6-P3", "Inbox/outbox test exists", has_inbox_test)
```

```python
# NEGATIVE SPACE (P16) -- Developer tools gaps

# Probe: pre-push inspector hook exists
pre_push = BROWSER_ROOT / "src" / "hooks" / "pre-push-inspector.sh"
has_hook = pre_push.exists() and pre_push.stat().st_size > 50
record("NS-P1", "Pre-push inspector hook exists", has_hook)

# Probe: CLI module exists for developer commands
cli_code = read_file(SRC / "cli" / "cli.py")
has_cli = bool(re.search(r'command|cli|argument|parse|run', cli_code, re.IGNORECASE))
record("NS-P2", "CLI module with developer commands exists", has_cli,
       f"{len(cli_code)} bytes")

# Probe: init workspace for project setup
init_ws = read_file(SRC / "cli" / "init_workspace.py")
has_init = bool(re.search(r'init|workspace|setup|scaffold|create', init_ws, re.IGNORECASE))
record("NS-P3", "Init workspace module for project setup exists", has_init)

# Probe: credential vault for secure dev auth
cred_vault = read_file(SRC / "cli" / "credential_vault.py")
has_cred = len(cred_vault) > 100
record("NS-P4", "Credential vault for dev authentication exists", has_cred,
       f"{len(cred_vault)} bytes")
```

```python
# EVIDENCE SUMMARY -- Tower 42 Developer Tools QA
passed = len([r for r in results if r["status"] == "PASS"])
failed = len([r for r in results if r["status"] == "FAIL"])
total = len(results)
score = round(passed / total * 100, 1) if total > 0 else 0
finding_rate = round(failed / total * 100, 1) if total > 0 else 0
grade = "A+" if score >= 95 else "A" if score >= 90 else "B+" if score >= 85 else "B" if score >= 80 else "C" if score >= 70 else "D" if score >= 60 else "F"

summary = {
    "surface": NORTHSTAR, "tower": TOWER, "tower_name": TOWER_NAME,
    "timestamp": datetime.now().isoformat(),
    "total_probes": total, "passed": passed, "failed": failed,
    "score": score, "grade": grade, "finding_rate": finding_rate,
    "converged": finding_rate < 5,
    "findings": [r for r in results if r["status"] == "FAIL"],
    "evidence_hash": hashlib.sha256(json.dumps(results, sort_keys=True).encode()).hexdigest()[:16]
}

print("=" * 60)
print(f"TOWER {TOWER}: {TOWER_NAME}")
print("=" * 60)
print(f"Probes: {total} | Passed: {passed} | Failed: {failed}")
print(f"Score: {score}% | Grade: {grade} | Finding rate: {finding_rate}%")
print(f"Converged: {'YES' if summary['converged'] else 'NO'}")
print(f"Evidence hash: {summary['evidence_hash']}")
if summary["findings"]:
    print("\nFINDINGS:")
    for f in summary["findings"]:
        print(f"  [{f['id']}] {f['name']}: {f.get('detail', '')}")
print()
print(json.dumps(summary, indent=2))
```
