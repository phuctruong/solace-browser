```python
# NORTHSTAR -- Tower 43 Sales & Marketing QA for Solace Browser
# DNA: sales(pipeline) = leads(track) x prices(monitor) x outreach(automate) x linkedin(optimize) x nurture(sequence) x convert(close) x roi(measure)
# Probes marketing pages, CTAs, SEO elements, conversion flow, analytics refs
import os, sys, re, json, hashlib
from pathlib import Path
from datetime import datetime

NORTHSTAR = "sales-marketing-t43-qa"
NOTEBOOK_PATH = "notebooks/qa/61-sales-marketing-t43-qa.ipynb"
TOWER = 43
TOWER_NAME = "Tower of Sales & Marketing"
MIN_SCORE = 70

BROWSER_ROOT = Path("/home/phuc/projects/solace-browser")
SRC = BROWSER_ROOT / "src"
WEB = BROWSER_ROOT / "web"
TESTS = BROWSER_ROOT / "tests"
DATA = BROWSER_ROOT / "data" / "default"

results = []

def record(probe_id, name, passed, detail=""):
    status = "PASS" if passed else "FAIL"
    results.append({"id": probe_id, "name": name, "status": status, "detail": detail})
    print(f"{status}: {name}" + (f" -- {detail}" if detail else ""))

def read_file(path):
    p = Path(path)
    return p.read_text(encoding='utf-8', errors='replace') if p.exists() else ""

print(f"Tower {TOWER}: {TOWER_NAME}")
assert BROWSER_ROOT.exists()
```

```python
# F1 -- Vanessa Liu: Lead Tracking
# Lead pipeline recipe must exist

# Probe: Lead pipeline recipe exists
lead_recipe = read_file(DATA / "apps" / "lead-pipeline" / "recipe.json")
has_lead = len(lead_recipe) > 50
record("F1-P1", "Lead pipeline recipe exists", has_lead,
       f"{len(lead_recipe)} bytes")

# Probe: Lead pipeline budget exists
lead_budget = read_file(DATA / "apps" / "lead-pipeline" / "budget.json")
has_lead_budget = len(lead_budget) > 20
record("F1-P2", "Lead pipeline budget.json exists", has_lead_budget,
       f"{len(lead_budget)} bytes")

# F2 -- Carlos Ramirez: Price Monitoring
# Amazon price tracker recipe exists
price_recipe = read_file(DATA / "apps" / "amazon-price-tracker" / "recipe.json")
has_price = len(price_recipe) > 50
record("F2-P1", "Amazon price tracker recipe exists", has_price,
       f"{len(price_recipe)} bytes")

# Probe: Amazon price tracker budget exists
price_budget = read_file(DATA / "apps" / "amazon-price-tracker" / "budget.json")
has_price_budget = len(price_budget) > 20
record("F2-P2", "Amazon price tracker budget.json exists", has_price_budget,
       f"{len(price_budget)} bytes")
```

```python
# F3 -- Sarah Mitchell: Outreach Automation
# LinkedIn outreach recipe for cold outreach

# Probe: LinkedIn outreach recipe exists
outreach_recipe = read_file(DATA / "apps" / "linkedin-outreach" / "recipe.json")
has_outreach = len(outreach_recipe) > 50
record("F3-P1", "LinkedIn outreach recipe exists", has_outreach,
       f"{len(outreach_recipe)} bytes")

# F4 -- David Chung: LinkedIn Optimization
# LinkedIn profile optimizer recipe exists
li_optimizer = read_file(DATA / "apps" / "linkedin-profile-optimizer" / "recipe.json")
has_optimizer = len(li_optimizer) > 50
record("F4-P1", "LinkedIn profile optimizer recipe exists", has_optimizer,
       f"{len(li_optimizer)} bytes")

# Probe: LinkedIn scopes support profile reading
scopes_code = read_file(SRC / "oauth3" / "scopes.py")
has_li_profile = bool(re.search(r'linkedin\.read\.profile|linkedin\.read_profile', scopes_code))
record("F4-P2", "LinkedIn profile read scope defined", has_li_profile)

# Probe: Google search trends recipe (market research)
trends_recipe = read_file(DATA / "apps" / "google-search-trends" / "recipe.json")
has_trends = len(trends_recipe) > 50
record("F4-P3", "Google search trends recipe exists", has_trends,
       f"{len(trends_recipe)} bytes")
```

```python
# F5/F6 -- SEO and Marketing Pages Quality
# OG tags, meta descriptions, canonical links, CTAs on key pages

# Probe: Home page has CTA elements
home_html = read_file(WEB / "home.html")
has_home_cta = bool(re.search(r'get.started|sign.up|try.free|download|start', home_html, re.IGNORECASE))
record("F5-P1", "Home page has CTA elements", has_home_cta)

# Probe: App store page exists and has content
appstore_html = read_file(WEB / "app-store.html")
has_appstore = len(appstore_html) > 500
record("F5-P2", "App store page exists with content", has_appstore,
       f"{len(appstore_html)} bytes")

# Probe: robots.txt exists for SEO
robots = read_file(WEB / "robots.txt")
has_robots = len(robots) > 10
record("F5-P3", "robots.txt exists", has_robots,
       f"{len(robots)} bytes")

# Probe: sitemap.xml exists for SEO
sitemap = read_file(WEB / "sitemap.xml")
has_sitemap = len(sitemap) > 50
record("F5-P4", "sitemap.xml exists", has_sitemap,
       f"{len(sitemap)} bytes")

# Probe: OG tags on key pages
html_files = list(WEB.glob("*.html"))
og_count = 0
for f in html_files:
    content = f.read_text(encoding='utf-8', errors='replace')
    if re.search(r'og:title', content):
        og_count += 1
record("F6-P1", "OG tags present on multiple pages (>=3)",
       og_count >= 3, f"{og_count} pages with og:title")

# Probe: Meta descriptions on pages
desc_count = 0
for f in html_files:
    content = f.read_text(encoding='utf-8', errors='replace')
    if re.search(r'meta.*name=["\']description', content):
        desc_count += 1
record("F6-P2", "Meta descriptions on multiple pages (>=3)",
       desc_count >= 3, f"{desc_count} pages with meta description")
```

```python
# F7 -- Jennifer Okafor: ROI Measurement
# Budget tracking, cost measurement infrastructure

# Probe: Budget gates for cost enforcement
budget_gates = read_file(SRC / "budget_gates.py")
has_budget = bool(re.search(r'budget|cost|limit|cap|spend', budget_gates, re.IGNORECASE))
record("F7-P1", "Budget gates module for cost tracking exists", has_budget,
       f"{len(budget_gates)} bytes")

# Probe: Store client metrics reporter for ROI data
metrics_reporter = read_file(SRC / "store_client" / "metrics_reporter.py")
has_reporter = bool(re.search(r'metric|report|usage|cost', metrics_reporter, re.IGNORECASE))
record("F7-P2", "Metrics reporter for ROI data exists", has_reporter,
       f"{len(metrics_reporter)} bytes")

# Probe: Download page has clear pricing/value proposition
download_html = read_file(WEB / "download.html")
has_download = len(download_html) > 500
record("F7-P3", "Download page exists with value proposition", has_download,
       f"{len(download_html)} bytes")

# Probe: Competitor watch recipe (competitive intelligence)
comp_recipe = read_file(DATA / "apps" / "competitor-watch" / "recipe.json")
has_comp = len(comp_recipe) > 50
record("F7-P4", "Competitor watch recipe exists", has_comp,
       f"{len(comp_recipe)} bytes")
```

```python
# NEGATIVE SPACE (P16) -- Marketing gaps

# Probe: Canonical URLs on pages
html_files = list(WEB.glob("*.html"))
canon_count = 0
for f in html_files:
    content = f.read_text(encoding='utf-8', errors='replace')
    if re.search(r'rel=["\']canonical', content):
        canon_count += 1
record("NS-P1", "Canonical URLs present on multiple pages (>=2)",
       canon_count >= 2, f"{canon_count} pages with canonical")

# Probe: Twitter card meta on pages
twitter_count = 0
for f in html_files:
    content = f.read_text(encoding='utf-8', errors='replace')
    if re.search(r'twitter:card', content):
        twitter_count += 1
record("NS-P2", "Twitter card meta on pages (>=2)",
       twitter_count >= 2, f"{twitter_count} pages with twitter:card")

# Probe: Guide page for onboarding/conversion
guide_html = read_file(WEB / "guide.html")
has_guide = len(guide_html) > 200
record("NS-P3", "Guide page exists for onboarding", has_guide,
       f"{len(guide_html)} bytes")

# Probe: App detail page for deeper engagement
detail_html = read_file(WEB / "app-detail.html")
has_detail = len(detail_html) > 500
record("NS-P4", "App detail page exists for deeper engagement", has_detail,
       f"{len(detail_html)} bytes")
```

```python
# EVIDENCE SUMMARY -- Tower 43 Sales & Marketing QA
passed = len([r for r in results if r["status"] == "PASS"])
failed = len([r for r in results if r["status"] == "FAIL"])
total = len(results)
score = round(passed / total * 100, 1) if total > 0 else 0
finding_rate = round(failed / total * 100, 1) if total > 0 else 0
grade = "A+" if score >= 95 else "A" if score >= 90 else "B+" if score >= 85 else "B" if score >= 80 else "C" if score >= 70 else "D" if score >= 60 else "F"

summary = {
    "surface": NORTHSTAR, "tower": TOWER, "tower_name": TOWER_NAME,
    "timestamp": datetime.now().isoformat(),
    "total_probes": total, "passed": passed, "failed": failed,
    "score": score, "grade": grade, "finding_rate": finding_rate,
    "converged": finding_rate < 5,
    "findings": [r for r in results if r["status"] == "FAIL"],
    "evidence_hash": hashlib.sha256(json.dumps(results, sort_keys=True).encode()).hexdigest()[:16]
}

print("=" * 60)
print(f"TOWER {TOWER}: {TOWER_NAME}")
print("=" * 60)
print(f"Probes: {total} | Passed: {passed} | Failed: {failed}")
print(f"Score: {score}% | Grade: {grade} | Finding rate: {finding_rate}%")
print(f"Converged: {'YES' if summary['converged'] else 'NO'}")
print(f"Evidence hash: {summary['evidence_hash']}")
if summary["findings"]:
    print("\nFINDINGS:")
    for f in summary["findings"]:
        print(f"  [{f['id']}] {f['name']}: {f.get('detail', '')}")
print()
print(json.dumps(summary, indent=2))
```
