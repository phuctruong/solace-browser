```python
import os, sys, re, json, hashlib
from pathlib import Path
from datetime import datetime

NORTHSTAR = "competitor-intel-t37-qa"
NOTEBOOK_PATH = "notebooks/qa/56-competitor-intel-t37-qa.ipynb"
TOWER = 37
TOWER_NAME = "Tower of Competitor Intelligence"
MIN_SCORE = 70
BROWSER_ROOT = Path("/home/phuc/projects/solace-browser")
SRC = BROWSER_ROOT / "src"
WEB = BROWSER_ROOT / "web"
TESTS = BROWSER_ROOT / "tests"
results = []
def record(probe_id, name, passed, detail=""):
    status = "PASS" if passed else "FAIL"
    results.append({"id": probe_id, "name": name, "status": status, "detail": detail})
    print(f"{status}: {name}" + (f" -- {detail}" if detail else ""))
def read_file(path):
    p = Path(path)
    return p.read_text(encoding='utf-8', errors='replace') if p.exists() else ""
print(f"Tower {TOWER}: {TOWER_NAME}")
assert BROWSER_ROOT.exists()
```

```python
# Cell 1: Competitor watch recipe structure
print("=== Competitor Watch Recipe ===")

apps_dir = BROWSER_ROOT / "data" / "default" / "apps"
cw_recipe = read_file(apps_dir / "competitor-watch" / "recipe.json")
cw_data = json.loads(cw_recipe) if cw_recipe else {}

# P01: Competitor watch recipe exists
record("T37-P01", "Competitor watch recipe exists",
       '"id": "competitor-watch"' in cw_recipe,
       "competitor-watch/recipe.json")

# P02: Recipe has google.read.search scope
record("T37-P02", "Competitor watch uses google.read.search scope",
       '"google.read.search"' in cw_recipe,
       "Scoped to read-only Google Search")

# P03: Recipe tracks named competitors (OpenAI, Anthropic, Bardeen, Browser-Use, Copilot, OpenClaw)
competitors = cw_data.get("input_params", {}).get("competitors", {}).get("default", [])
record("T37-P03", "Recipe defines default competitor list (6 competitors)",
       len(competitors) >= 5,
       f"Competitors: {competitors}")

# P04: Recipe loops over competitors for individual searches
cw_steps = json.dumps(cw_data.get("steps", []))
record("T37-P04", "Recipe loops over competitors for individual searches",
       "loop" in cw_steps and "competitor" in cw_steps,
       "Step 2 loops over params.competitors")

# P05: Recipe searches across TechCrunch, The Verge, Hacker News
record("T37-P05", "Recipe searches across TechCrunch, The Verge, Hacker News",
       "techcrunch.com" in cw_steps and "theverge.com" in cw_steps and "ycombinator.com" in cw_steps,
       "Multi-source competitor news scanning")
```

```python
# Cell 2: Competitor analysis and report generation
print("=== Analysis & Report Generation ===")

cw_recipe = read_file(BROWSER_ROOT / "data" / "default" / "apps" / "competitor-watch" / "recipe.json")
cw_data = json.loads(cw_recipe) if cw_recipe else {}
cw_steps = json.dumps(cw_data.get("steps", []))

# P06: Recipe uses llm_analyze for competitive analysis
record("T37-P06", "Recipe uses llm_analyze for competitive intelligence analysis",
       "llm_analyze" in cw_steps and "competitive intelligence" in cw_steps.lower(),
       "LLM analyzes signals for threat level and actionable response")

# P07: Analysis prompt covers threat levels (HIGH, MEDIUM, LOW)
record("T37-P07", "LLM prompt defines threat levels (HIGH, MEDIUM, LOW)",
       "HIGH" in cw_steps and "MEDIUM" in cw_steps and "LOW" in cw_steps,
       "Threat classification: HIGH (direct overlap), MEDIUM (adjacent), LOW (different market)")

# P08: Analysis identifies Solace advantages (OAuth3, local-first, evidence, recipe replay)
record("T37-P08", "LLM prompt identifies Solace competitive advantages",
       "OAuth3" in cw_steps and "local-first" in cw_steps and "evidence" in cw_steps,
       "Prompt explicitly frames Solace advantages vs each competitor")

# P09: Recipe generates markdown report with action items
record("T37-P09", "Recipe generates markdown report with action items section",
       "generate_report" in cw_steps and "action_items" in cw_steps,
       "Report template includes ## Action Items section")

# P10: Report is saved to outbox
record("T37-P10", "Competitive report saved to outbox with dated filename",
       "save_to_outbox" in cw_steps and "competitor-watch-{date}" in cw_steps,
       "Output: competitor-watch-{date}.json")
```

```python
# Cell 3: Google Search trends recipe for broader intel
print("=== Google Search Trends (Broader Intel) ===")

trends_recipe = read_file(BROWSER_ROOT / "data" / "default" / "apps" / "google-search-trends" / "recipe.json")
trends_data = json.loads(trends_recipe) if trends_recipe else {}
trends_steps = json.dumps(trends_data.get("steps", []))

# P11: Trends recipe scans Google News tech section
record("T37-P11", "Trends recipe scans Google News technology section",
       "news.google.com" in trends_steps,
       "Step 1 navigates to Google News Technology section")

# P12: Trends recipe also scans Hacker News front page
record("T37-P12", "Trends recipe scans Hacker News front page",
       "news.ycombinator.com" in trends_steps,
       "Step 3 navigates to HN for additional signal")

# P13: Trends recipe extracts structured fields (headline, source, link, time)
record("T37-P13", "Trends recipe extracts headline, source, link, time",
       "headline" in trends_steps and "source" in trends_steps and "link" in trends_steps,
       "Structured extraction from Google News articles")

# P14: Trends LLM analysis mentions competitor names
record("T37-P14", "Trends LLM prompt references competitor names for positioning",
       "Copilot" in trends_steps and "Bardeen" in trends_steps and "Browser-Use" in trends_steps,
       "Competitor mentions in trend analysis prompt")

# P15: Trends recipe has budget constraints
trends_budget = trends_data.get("budget", {})
record("T37-P15", "Trends recipe has budget with cost cap",
       trends_budget.get("max_cost_per_run_usd", 0) > 0,
       f"max_cost_per_run_usd={trends_budget.get('max_cost_per_run_usd')}")
```

```python
# Cell 4: Infrastructure for intel data collection and evidence
print("=== Intel Infrastructure ===")

executor_py = read_file(SRC / "recipes" / "recipe_executor.py")

# P16: Recipe executor supports extract action for web data collection
record("T37-P16", "Recipe executor supports extract action for structured data collection",
       '"extract"' in executor_py and "fields_requested" in executor_py,
       "extract action pulls structured fields from DOM selectors")

# P17: Recipe executor supports search action
record("T37-P17", "Recipe executor supports search action for keyword queries",
       '"search"' in executor_py and "query" in executor_py,
       "search action supports query parameter for keyword searches")

# P18: Recipe executor supports generate_report action
record("T37-P18", "Recipe executor supports generate_report action for intel reports",
       '"generate_report"' in executor_py and "output_format" in executor_py,
       "generate_report creates formatted markdown/text reports")

# P19: Competitive positioning diagram exists in src/diagrams
comp_diagram = BROWSER_ROOT / "src" / "diagrams" / "15-competitive-position.md"
diagram_content = read_file(comp_diagram)
record("T37-P19", "Competitive positioning diagram exists in src/diagrams",
       comp_diagram.exists() and len(diagram_content) > 100,
       "15-competitive-position.md with market positioning")

# P20: Reddit scanner recipe exists for social signal collection
reddit_recipe = read_file(BROWSER_ROOT / "data" / "default" / "apps" / "reddit-scanner" / "recipe.json")
record("T37-P20", "Reddit scanner recipe exists for social signal collection",
       '"reddit-scanner"' in reddit_recipe,
       "reddit-scanner provides additional competitive signal source")
```

```python
# Cell 5: Evidence summary
passed = len([r for r in results if r["status"] == "PASS"])
failed = len([r for r in results if r["status"] == "FAIL"])
total = len(results)
score = round(passed / total * 100, 1) if total > 0 else 0
finding_rate = round(failed / total * 100, 1) if total > 0 else 0
grade = "A+" if score >= 95 else "A" if score >= 90 else "B+" if score >= 85 else "B" if score >= 80 else "C" if score >= 70 else "D" if score >= 60 else "F"
summary = {"surface": NORTHSTAR, "tower": TOWER, "tower_name": TOWER_NAME, "timestamp": datetime.now().isoformat(), "total_probes": total, "passed": passed, "failed": failed, "score": score, "grade": grade, "finding_rate": finding_rate, "converged": finding_rate < 5, "findings": [r for r in results if r["status"] == "FAIL"], "evidence_hash": hashlib.sha256(json.dumps(results, sort_keys=True).encode()).hexdigest()[:16]}
print("=" * 60)
print(f"TOWER {TOWER}: {TOWER_NAME}")
print("=" * 60)
print(f"Probes: {total} | Passed: {passed} | Failed: {failed}")
print(f"Score: {score}% | Grade: {grade} | Finding rate: {finding_rate}%")
print(f"Converged: {'YES' if summary['converged'] else 'NO'}")
print(f"Evidence hash: {summary['evidence_hash']}")
if summary["findings"]:
    print("\nFINDINGS:")
    for f in summary["findings"]:
        print(f"  [{f['id']}] {f['name']}: {f.get('detail', '')}")
print()
print(json.dumps(summary, indent=2))
```
