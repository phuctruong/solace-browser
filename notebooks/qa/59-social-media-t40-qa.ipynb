```python
# NORTHSTAR -- Tower 40 Social Media Suite QA for Solace Browser
# DNA: social(suite) = post(multi) x schedule(auto) x monitor(brand) x engage(reply) x analytics(track) x content(generate) x budget(cap)
# Probes social media recipes, platform scopes, sharing features, community links
import os, sys, re, json, hashlib
from pathlib import Path
from datetime import datetime

NORTHSTAR = "social-media-t40-qa"
NOTEBOOK_PATH = "notebooks/qa/59-social-media-t40-qa.ipynb"
TOWER = 40
TOWER_NAME = "Tower of Social Media Suite"
MIN_SCORE = 70

BROWSER_ROOT = Path("/home/phuc/projects/solace-browser")
SRC = BROWSER_ROOT / "src"
WEB = BROWSER_ROOT / "web"
TESTS = BROWSER_ROOT / "tests"
DATA = BROWSER_ROOT / "data" / "default"

results = []

def record(probe_id, name, passed, detail=""):
    status = "PASS" if passed else "FAIL"
    results.append({"id": probe_id, "name": name, "status": status, "detail": detail})
    print(f"{status}: {name}" + (f" -- {detail}" if detail else ""))

def read_file(path):
    p = Path(path)
    return p.read_text(encoding='utf-8', errors='replace') if p.exists() else ""

print(f"Tower {TOWER}: {TOWER_NAME}")
assert BROWSER_ROOT.exists()
```

```python
# F1 -- Jasmine Torres: Multi-Platform Posting
# Social media recipes must exist for LinkedIn, Twitter, Instagram, Reddit

# Probe: LinkedIn poster recipe exists with valid structure
linkedin_recipe = read_file(DATA / "apps" / "linkedin-poster" / "recipe.json")
has_linkedin = len(linkedin_recipe) > 50
record("F1-P1", "LinkedIn poster recipe exists", has_linkedin,
       f"{len(linkedin_recipe)} bytes")

# Probe: Twitter poster recipe exists
twitter_recipe = read_file(DATA / "apps" / "twitter-poster" / "recipe.json")
has_twitter = len(twitter_recipe) > 50
record("F1-P2", "Twitter poster recipe exists", has_twitter,
       f"{len(twitter_recipe)} bytes")

# Probe: Instagram poster recipe exists
instagram_recipe = read_file(DATA / "apps" / "instagram-poster" / "recipe.json")
has_instagram = len(instagram_recipe) > 50
record("F1-P3", "Instagram poster recipe exists", has_instagram,
       f"{len(instagram_recipe)} bytes")

# Probe: Reddit scanner recipe exists
reddit_recipe = read_file(DATA / "apps" / "reddit-scanner" / "recipe.json")
has_reddit = len(reddit_recipe) > 50
record("F1-P4", "Reddit scanner recipe exists", has_reddit,
       f"{len(reddit_recipe)} bytes")
```

```python
# F2 -- Andrew Kim: Content Scheduling
# Schedule system with calendar view, timezone support

# Probe: Schedule calendar JS exists
schedule_cal = read_file(WEB / "js" / "schedule-calendar.js")
has_cal = bool(re.search(r'calendar|week|day|month|drag', schedule_cal, re.IGNORECASE))
record("F2-P1", "Schedule calendar JS module exists", has_cal,
       f"{len(schedule_cal)} bytes")

# Probe: Schedule core JS exists
schedule_core = read_file(WEB / "js" / "schedule-core.js")
has_core = bool(re.search(r'schedule|queue|run|task|time', schedule_core, re.IGNORECASE))
record("F2-P2", "Schedule core JS module exists", has_core,
       f"{len(schedule_core)} bytes")

# Probe: Schedule HTML page exists
schedule_html = read_file(WEB / "schedule.html")
has_schedule_page = len(schedule_html) > 200
record("F2-P3", "Schedule page exists", has_schedule_page,
       f"{len(schedule_html)} bytes")

# Probe: Schedule CSS exists
schedule_css = read_file(WEB / "css" / "schedule.css")
has_schedule_css = len(schedule_css) > 100
record("F2-P4", "Schedule CSS exists", has_schedule_css,
       f"{len(schedule_css)} bytes")
```

```python
# F3 -- Rachel Okonkwo: Brand Monitoring
# Twitter monitor recipe + competitor watch recipe

# Probe: Twitter monitor recipe exists
tw_monitor = read_file(DATA / "apps" / "twitter-monitor" / "recipe.json")
has_tw_monitor = len(tw_monitor) > 50
record("F3-P1", "Twitter monitor recipe exists", has_tw_monitor,
       f"{len(tw_monitor)} bytes")

# Probe: Competitor watch recipe exists
comp_watch = read_file(DATA / "apps" / "competitor-watch" / "recipe.json")
has_comp = len(comp_watch) > 50
record("F3-P2", "Competitor watch recipe exists", has_comp,
       f"{len(comp_watch)} bytes")

# F4 -- Lucas Petersen: Engagement Tracking
# LinkedIn outreach recipe exists
li_outreach = read_file(DATA / "apps" / "linkedin-outreach" / "recipe.json")
has_outreach = len(li_outreach) > 50
record("F4-P1", "LinkedIn outreach recipe exists", has_outreach,
       f"{len(li_outreach)} bytes")

# Probe: LinkedIn platform scopes defined in OAuth3
scopes_code = read_file(SRC / "oauth3" / "scopes.py")
linkedin_scopes = re.findall(r'"linkedin\.[\w.]+"', scopes_code)
record("F4-P2", "LinkedIn OAuth3 scopes defined (>=5)",
       len(linkedin_scopes) >= 5,
       f"{len(linkedin_scopes)} LinkedIn scopes")

# Probe: Reddit platform scopes defined
reddit_scopes = re.findall(r'"reddit\.[\w.]+"', scopes_code)
record("F4-P3", "Reddit OAuth3 scopes defined (>=3)",
       len(reddit_scopes) >= 3,
       f"{len(reddit_scopes)} Reddit scopes")
```

```python
# F5 -- Samira Al-Hassan: Content Generation
# YouTube script writer recipe + LLM clients for AI content generation

# Probe: YouTube script writer recipe exists
yt_recipe = read_file(DATA / "apps" / "youtube-script-writer" / "recipe.json")
has_yt = len(yt_recipe) > 50
record("F5-P1", "YouTube script writer recipe exists", has_yt,
       f"{len(yt_recipe)} bytes")

# Probe: LLM client for AI content generation
claude_client = read_file(SRC / "llm" / "claude_code_client.py")
together_client = read_file(SRC / "llm" / "together_client.py")
has_llm = len(claude_client) > 100 or len(together_client) > 100
record("F5-P2", "LLM clients exist for AI content generation", has_llm)

# F6 -- Brian Nakamura: Analytics Dashboard
# App store page shows apps by category (Social category exists)
app_store_html = read_file(WEB / "app-store.html")
has_social_category = bool(re.search(r'social|Social', app_store_html))
record("F6-P1", "App store mentions Social category", has_social_category)

# F7 -- Eleanor Vasquez: Budget Enforcement
# Budget files exist for social media apps
li_budget = read_file(DATA / "apps" / "linkedin-poster" / "budget.json")
has_budget = len(li_budget) > 20
record("F7-P1", "LinkedIn poster budget.json exists", has_budget,
       f"{len(li_budget)} bytes")

tw_budget = read_file(DATA / "apps" / "twitter-poster" / "budget.json")
has_tw_budget = len(tw_budget) > 20
record("F7-P2", "Twitter poster budget.json exists", has_tw_budget,
       f"{len(tw_budget)} bytes")

# Probe: Budget gates module exists
budget_gates = read_file(SRC / "budget_gates.py")
has_gates = bool(re.search(r'budget|limit|cap|enforce|gate', budget_gates, re.IGNORECASE))
record("F7-P3", "Budget gates enforcement module exists", has_gates,
       f"{len(budget_gates)} bytes")
```

```python
# NEGATIVE SPACE (P16) -- Social media infrastructure gaps

# Probe: Multi-platform recipe validation tests exist
multi_test = TESTS / "test_multi_platform_recipe_validation.py"
has_multi_test = multi_test.exists() and multi_test.stat().st_size > 100
record("NS-P1", "Multi-platform recipe validation test exists", has_multi_test)

# Probe: Schedule cloud module exists for cloud-synced scheduling
schedule_cloud = read_file(WEB / "js" / "schedule-cloud.js")
has_cloud_sched = bool(re.search(r'cloud|sync|upload|push', schedule_cloud, re.IGNORECASE))
record("NS-P2", "Schedule cloud sync JS module exists", has_cloud_sched,
       f"{len(schedule_cloud)} bytes")

# Probe: Schedule evidence module exists
schedule_evidence = read_file(WEB / "js" / "schedule-evidence.js")
has_sched_evidence = bool(re.search(r'evidence|hash|seal|chain', schedule_evidence, re.IGNORECASE))
record("NS-P3", "Schedule evidence JS module exists", has_sched_evidence,
       f"{len(schedule_evidence)} bytes")

# Probe: Schedule approvals module exists
schedule_approvals = read_file(WEB / "js" / "schedule-approvals.js")
has_approvals = bool(re.search(r'approv|confirm|consent|gate', schedule_approvals, re.IGNORECASE))
record("NS-P4", "Schedule approvals JS module exists", has_approvals,
       f"{len(schedule_approvals)} bytes")
```

```python
# EVIDENCE SUMMARY -- Tower 40 Social Media Suite QA
passed = len([r for r in results if r["status"] == "PASS"])
failed = len([r for r in results if r["status"] == "FAIL"])
total = len(results)
score = round(passed / total * 100, 1) if total > 0 else 0
finding_rate = round(failed / total * 100, 1) if total > 0 else 0
grade = "A+" if score >= 95 else "A" if score >= 90 else "B+" if score >= 85 else "B" if score >= 80 else "C" if score >= 70 else "D" if score >= 60 else "F"

summary = {
    "surface": NORTHSTAR, "tower": TOWER, "tower_name": TOWER_NAME,
    "timestamp": datetime.now().isoformat(),
    "total_probes": total, "passed": passed, "failed": failed,
    "score": score, "grade": grade, "finding_rate": finding_rate,
    "converged": finding_rate < 5,
    "findings": [r for r in results if r["status"] == "FAIL"],
    "evidence_hash": hashlib.sha256(json.dumps(results, sort_keys=True).encode()).hexdigest()[:16]
}

print("=" * 60)
print(f"TOWER {TOWER}: {TOWER_NAME}")
print("=" * 60)
print(f"Probes: {total} | Passed: {passed} | Failed: {failed}")
print(f"Score: {score}% | Grade: {grade} | Finding rate: {finding_rate}%")
print(f"Converged: {'YES' if summary['converged'] else 'NO'}")
print(f"Evidence hash: {summary['evidence_hash']}")
if summary["findings"]:
    print("\nFINDINGS:")
    for f in summary["findings"]:
        print(f"  [{f['id']}] {f['name']}: {f.get('detail', '')}")
print()
print(json.dumps(summary, indent=2))
```
