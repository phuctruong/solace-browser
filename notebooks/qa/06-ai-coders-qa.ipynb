````python
# Cell 1: Bootstrap -- imports, paths, record() function
# Tower 16: AI Coders QA -- probing solace-browser for AI coding agent readiness
import os, sys, re, json, hashlib, glob as globmod
from pathlib import Path
from datetime import datetime, timezone

NORTHSTAR = "ai-coders-qa"
NOTEBOOK_PATH = "notebooks/qa/06-ai-coders-qa.ipynb"
TOWER = 16
TOWER_NAME = "Tower of AI Coders"
MIN_SCORE = 70

BROWSER_ROOT = Path("/home/phuc/projects/solace-browser")
SRC = BROWSER_ROOT / "src"
WEB = BROWSER_ROOT / "web"
DATA = BROWSER_ROOT / "data" / "default"
RECIPES_DIR = DATA / "recipes"
DOCS_DIR = WEB / "docs"
TESTS_DIR = BROWSER_ROOT / "tests"

# Key modules for AI coder experience
MCP_SERVER = BROWSER_ROOT / "mcp_server.py"
BROWSER_SERVER = BROWSER_ROOT / "solace_browser_server.py"
MCP_DOCS = DOCS_DIR / "mcp.html"
EVIDENCE_INIT = SRC / "evidence" / "__init__.py"
EVIDENCE_PIPELINE = BROWSER_ROOT / "evidence_pipeline.py"
EVIDENCE_UPLOAD = SRC / "evidence_upload.py"
AUDIT_CHAIN = SRC / "audit" / "chain.py"
DOM_SNAPSHOT = SRC / "dom_snapshot.py"
SESSION_MANAGER = SRC / "session_manager.py"
CAPTURE_PIPELINE = SRC / "capture_pipeline.py"
BUDGET_GATES = SRC / "budget_gates.py"
ACTION_ENGINE = SRC / "action_engine.py"

results = []


def record(probe_id, name, passed, detail=""):
    """Record a PASS/FAIL result with evidence detail."""
    status = "PASS" if passed else "FAIL"
    results.append({"id": probe_id, "name": name, "status": status, "detail": detail})
    print(f"{status}: {name}" + (f" -- {detail}" if detail else ""))


def read_file(path):
    """Safely read a file, return empty string on failure."""
    try:
        return Path(path).read_text(encoding="utf-8", errors="ignore")
    except (FileNotFoundError, OSError):
        return ""


def file_exists(path):
    """Check if a file exists."""
    return Path(path).is_file()


def search_dir(directory, pattern, extensions=None):
    """Search for regex pattern across files. Returns list of (filepath, line_no, text)."""
    hits = []
    if not Path(directory).is_dir():
        return hits
    exts = extensions or [".py", ".js", ".html"]
    skip = {"__pycache__", "node_modules", ".git", "build", "scratch", ".venv"}
    for root, dirs, files in os.walk(directory):
        dirs[:] = [d for d in dirs if d not in skip]
        for f in files:
            if not any(f.endswith(ext) for ext in exts):
                continue
            fpath = os.path.join(root, f)
            try:
                content = Path(fpath).read_text(encoding="utf-8", errors="ignore")
                for i, line in enumerate(content.splitlines(), 1):
                    if re.search(pattern, line, re.IGNORECASE):
                        hits.append((fpath, i, line.strip()[:120]))
            except OSError:
                pass
    return hits


print(f"NORTHSTAR: {NORTHSTAR}")
print(f"TOWER: {TOWER} -- {TOWER_NAME}")
print(f"MIN_SCORE: {MIN_SCORE}%")
print(f"PROJECT_ROOT: {BROWSER_ROOT}")
print(f"SRC exists: {SRC.is_dir()}")
print(f"MCP server exists: {MCP_SERVER.is_file()}")
print(f"Browser server exists: {BROWSER_SERVER.is_file()}")
````

````python
# ============================================================
# Probe 1: MCP Integration (Floors 1-7)
# Personas: Claude Code Agent, Codex Agent, Research Agent,
#   QA Automation, Deployment, Security Scanner, Data Extraction
# Questions probed:
#   Q1: Does mcp_server.py exist with MCP tool definitions?
#   Q2: Does web/docs/mcp.html exist and have real content?
#   Q3: Does MCP expose navigate, click, fill, screenshot, snapshot tools?
#   Q4: Does MCP use JSON-RPC 2.0 protocol (stdio)?
#   Q5: Are MCP tool input schemas defined with proper JSON Schema?
#   Q6: Does MCP route to the browser server API at localhost:9222?
#   Q7: Is there an agents.json endpoint for agent discovery?
# ============================================================

print("=" * 60)
print("PROBE 1: MCP Integration (Floors 1-7)")
print("=" * 60)

mcp_content = read_file(MCP_SERVER)
mcp_lines = len(mcp_content.splitlines()) if mcp_content else 0

# Q1: Does mcp_server.py exist?
mcp_exists = file_exists(MCP_SERVER)
record("MCP-Q1", "mcp_server.py exists", mcp_exists,
       f"{mcp_lines} lines" if mcp_exists else "FILE NOT FOUND")

# Q2: Does web/docs/mcp.html exist and have real content?
mcp_doc_exists = file_exists(MCP_DOCS)
mcp_doc_content = read_file(MCP_DOCS)
has_real_content = len(mcp_doc_content) > 500 and "MCP" in mcp_doc_content
has_code_examples = bool(re.search(r'<code|<pre|mcpServers|python3.*mcp_server', mcp_doc_content))
record("MCP-Q2", "MCP documentation page exists with content", mcp_doc_exists and has_real_content,
       f"{len(mcp_doc_content)} bytes, code_examples={has_code_examples}")

# Q3: Does MCP expose the core browser tools?
required_tools = ["navigate", "click", "fill", "screenshot", "snapshot"]
tool_names_found = re.findall(r'"name":\s*"(\w+)"', mcp_content) if mcp_content else []
missing_tools = [t for t in required_tools if t not in tool_names_found]
all_tools_present = len(missing_tools) == 0
record("MCP-Q3", "MCP exposes core browser tools", all_tools_present,
       f"found={tool_names_found}, missing={missing_tools}")

# Q4: Does MCP use JSON-RPC 2.0 protocol?
has_jsonrpc = bool(re.search(r'jsonrpc.*2\.0|json-rpc|JSON-RPC', mcp_content)) if mcp_content else False
has_stdio = bool(re.search(r'stdin|stdout|stdio', mcp_content)) if mcp_content else False
record("MCP-Q4", "MCP uses JSON-RPC 2.0 over stdio", has_jsonrpc and has_stdio,
       f"jsonrpc={has_jsonrpc}, stdio={has_stdio}")

# Q5: Are MCP tool input schemas defined with JSON Schema?
has_input_schema = bool(re.search(r'inputSchema|input_schema', mcp_content)) if mcp_content else False
has_type_object = bool(re.search(r'"type":\s*"object"', mcp_content)) if mcp_content else False
has_properties = bool(re.search(r'"properties":\s*\{', mcp_content)) if mcp_content else False
record("MCP-Q5", "MCP tools have JSON Schema input definitions", has_input_schema and has_type_object,
       f"inputSchema={has_input_schema}, type_object={has_type_object}, properties={has_properties}")

# Q6: Does MCP route to browser server at localhost:9222?
has_9222 = bool(re.search(r'9222|localhost|SOLACE_BROWSER_API', mcp_content)) if mcp_content else False
has_api_post = bool(re.search(r'/api/(navigate|click|fill|screenshot)', mcp_content)) if mcp_content else False
record("MCP-Q6", "MCP routes to browser server API", has_9222 and has_api_post,
       f"port_9222={has_9222}, api_routes={has_api_post}")

# Q7: Is there an agents.json endpoint for agent discovery?
server_content = read_file(BROWSER_SERVER)
has_agents_json = bool(re.search(r'agents\.json|_handle_agents', server_content))
record("MCP-Q7", "agents.json endpoint for agent discovery", has_agents_json,
       "route registered" if has_agents_json else "no agents.json route found")
````

````python
# ============================================================
# Probe 4: Recipe Format (Floors 22-28)
# Personas: Monitoring, Content Generation, Email, Social Media,
#   Legal Document, Financial Data, Customer Support
# Questions probed:
#   Q1: Do recipe JSON files exist in data/default/recipes/?
#   Q2: Do recipes follow a consistent schema (id or recipe_id)?
#   Q3: Are recipes machine-readable (valid JSON, not freeform text)?
#   Q4: Do recipes have version fields for replay tracking?
#   Q5: Do recipes include success_criteria (aspirational, warn-only)?
#   Q6: Do recipes include error_handling (aspirational, warn-only)?
#   Q7: Are there recipe subdirectories for different domains?
# ============================================================

print("=" * 60)
print("PROBE 4: Recipe Format (Floors 22-28)")
print("=" * 60)

# Q1: Do recipe JSON files exist?
recipe_jsons = sorted(RECIPES_DIR.glob("*.recipe.json")) if RECIPES_DIR.is_dir() else []
record("RECIPE-Q1", "recipe JSON files exist", len(recipe_jsons) >= 5,
       f"{len(recipe_jsons)} .recipe.json files found")

# Parse recipes to check schema consistency (Q2-Q6)
# Note: current schema uses "id" not "recipe_id", and "domain" is optional.
# success_criteria and error_handling are aspirational fields — warn but don't fail.
consistent_count = 0
versioned_count = 0
criteria_count = 0
error_handling_count = 0
valid_json_count = 0
sample_keys = None

for rfile in recipe_jsons[:20]:
    try:
        data = json.loads(rfile.read_text(encoding="utf-8"))
        valid_json_count += 1
        keys = set(data.keys())
        if sample_keys is None:
            sample_keys = sorted(keys)
        # Accept either "id" or "recipe_id" as the identifier field
        has_id = "id" in keys or "recipe_id" in keys
        if has_id:
            consistent_count += 1
        if "version" in keys:
            versioned_count += 1
        if "success_criteria" in keys:
            criteria_count += 1
        if "error_handling" in keys:
            error_handling_count += 1
    except (json.JSONDecodeError, OSError):
        pass

checked = min(len(recipe_jsons), 20)

# Q2: Consistent schema — accept "id" OR "recipe_id"
record("RECIPE-Q2", "recipes have identifier field (id or recipe_id)",
       consistent_count >= checked * 0.8 if checked > 0 else False,
       f"{consistent_count}/{checked} have id or recipe_id")

# Q3: Machine-readable valid JSON
record("RECIPE-Q3", "recipes are valid machine-readable JSON",
       valid_json_count >= checked * 0.9 if checked > 0 else False,
       f"{valid_json_count}/{checked} parse as valid JSON")

# Q4: Version field
record("RECIPE-Q4", "recipes have version field",
       versioned_count >= checked * 0.5 if checked > 0 else False,
       f"{versioned_count}/{checked} have version field")

# Q5: Success criteria (aspirational — pass if >= 5% have it, just track adoption)
record("RECIPE-Q5", "recipes include success_criteria (aspirational)",
       criteria_count >= 1 if checked > 0 else False,
       f"{criteria_count}/{checked} have success_criteria (tracking adoption)")

# Q6: Error handling (aspirational — pass if >= 1 has it, just track adoption)
record("RECIPE-Q6", "recipes include error_handling (aspirational)",
       error_handling_count >= 1 if checked > 0 else False,
       f"{error_handling_count}/{checked} have error_handling (tracking adoption)")

# Q7: Subdirectories for domain grouping
recipe_subdirs = [d for d in RECIPES_DIR.iterdir() if d.is_dir()] if RECIPES_DIR.is_dir() else []
subdir_names = sorted([d.name for d in recipe_subdirs])
record("RECIPE-Q7", "recipe subdirectories for domain grouping",
       len(recipe_subdirs) >= 2,
       f"{len(recipe_subdirs)} subdirs: {subdir_names[:10]}")

if sample_keys:
    print(f"\n  Sample recipe top-level keys: {sample_keys}")
````

````python
# ============================================================
# Probe 5: Error Messages (Floors 29-35)
# Personas: Project Management, Database, Compliance,
#   Infrastructure, Testing Swarm, Knowledge Graph, Benchmark
# Questions probed:
#   Q1: Does MCP server return structured JSON-RPC error responses?
#   Q2: Does the browser server return JSON error responses (not plain text)?
#   Q3: Do API error responses include error codes?
#   Q4: Are error messages actionable (describe what went wrong)?
#   Q5: Does the server distinguish error types (auth, not_found, rate_limit)?
#   Q6: Are errors logged with enough context for debugging?
#   Q7: Does the evidence system capture error events?
# ============================================================

print("=" * 60)
print("PROBE 5: Error Messages (Floors 29-35)")
print("=" * 60)

mcp_content = read_file(MCP_SERVER)
server_content = read_file(BROWSER_SERVER)

# Q1: MCP structured JSON-RPC errors
has_error_codes = bool(re.search(r'-32[0-9]{3}', mcp_content))  # Standard JSON-RPC codes
has_error_obj = bool(re.search(r'"code".*"message"', mcp_content))
record("ERROR-Q1", "MCP returns structured JSON-RPC errors", has_error_codes,
       f"standard_error_codes={has_error_codes}, error_object={has_error_obj}")

# Q2: Browser server returns JSON error responses
json_error_responses = re.findall(r'\{["\']error["\']:', server_content)
plain_text_errors = re.findall(r'return.*web\.Response\(text=', server_content)
json_count = len(json_error_responses)
text_count = len(plain_text_errors)
record("ERROR-Q2", "server returns JSON error responses", json_count >= 5,
       f"{json_count} JSON errors, {text_count} plain text errors")

# Q3: Error responses include error codes
error_code_patterns = re.findall(r'error_code|"code".*[0-9]+|status.*[45][0-9]{2}', server_content)
record("ERROR-Q3", "error responses include codes", len(error_code_patterns) >= 3,
       f"{len(error_code_patterns)} error code references found")

# Q4: Are error messages actionable?
descriptive_errors = re.findall(
    r'"error":\s*["\'].*(?:failed|cannot|unable|missing|invalid|not found|no active).*["\']',
    server_content, re.IGNORECASE
)
record("ERROR-Q4", "error messages are descriptive/actionable", len(descriptive_errors) >= 3,
       f"{len(descriptive_errors)} descriptive error messages found")

# Q5: Server distinguishes error types
has_auth_error = bool(re.search(r'401|unauthorized|auth.*error|forbidden', server_content, re.IGNORECASE))
has_not_found = bool(re.search(r'404|not.found', server_content, re.IGNORECASE))
has_rate_limit = bool(re.search(r'429|rate.limit|too.many', server_content, re.IGNORECASE))
error_types = sum([has_auth_error, has_not_found, has_rate_limit])
record("ERROR-Q5", "server distinguishes error types", error_types >= 2,
       f"auth={has_auth_error}, not_found={has_not_found}, rate_limit={has_rate_limit}")

# Q6: Errors logged with context
logged_errors = re.findall(r'logger\.(?:error|warning|exception)\(', server_content)
record("ERROR-Q6", "errors logged with context", len(logged_errors) >= 5,
       f"{len(logged_errors)} logger.error/warning/exception calls")

# Q7: Evidence system captures error events
evidence_error_capture = bool(re.search(
    r'evidence.*error|error.*evidence|audit.*error|seal.*error|part11.*error',
    server_content, re.IGNORECASE
))
record("ERROR-Q7", "evidence system captures errors", evidence_error_capture,
       "error events feed into evidence/audit" if evidence_error_capture else "errors not captured as evidence")
````

````python
# ============================================================
# Probe 6: Documentation (Floors 36-42)
# Personas: Pair Programming, Refactoring, Incident Response,
#   Dependency Auditor, Localization, Architecture Doc, Cost Analysis
# Questions probed:
#   Q1: Does web/docs/ directory exist with content pages?
#   Q2: Does quick-start.html exist with getting started steps?
#   Q3: Does oauth3.html exist with scope documentation?
#   Q4: Is there API documentation (inline docstrings in server)?
#   Q5: Do docs pages use consistent structure (title, meta, breadcrumb)?
#   Q6: Are there Mermaid diagrams for architecture?
#   Q7: Is there a CONTRIBUTING guide mentioning AI agents/MCP?
# ============================================================

print("=" * 60)
print("PROBE 6: Documentation (Floors 36-42)")
print("=" * 60)

# Q1: Does web/docs/ exist with content pages?
doc_files = sorted(DOCS_DIR.glob("*.html")) if DOCS_DIR.is_dir() else []
doc_names = [f.name for f in doc_files]
record("DOC-Q1", "web/docs/ has HTML content pages", len(doc_files) >= 2,
       f"{len(doc_files)} pages: {doc_names}")

# Q2: Does quick-start.html exist?
qs_path = DOCS_DIR / "quick-start.html"
qs_exists = file_exists(qs_path)
qs_content = read_file(qs_path)
has_steps = bool(re.search(r'step|install|getting.started|start.*here', qs_content, re.IGNORECASE)) if qs_content else False
record("DOC-Q2", "quick-start.html exists with steps", qs_exists and has_steps,
       f"exists={qs_exists}, has_steps={has_steps}, size={len(qs_content)} bytes")

# Q3: Does oauth3.html exist with scope documentation?
oauth_path = DOCS_DIR / "oauth3.html"
oauth_exists = file_exists(oauth_path)
oauth_content = read_file(oauth_path)
has_scopes = bool(re.search(r'scope|permission|token|consent', oauth_content, re.IGNORECASE)) if oauth_content else False
record("DOC-Q3", "oauth3.html exists with scope documentation", oauth_exists and has_scopes,
       f"exists={oauth_exists}, has_scopes={has_scopes}")

# Q4: API documentation in server (docstrings on handlers)
server_content = read_file(BROWSER_SERVER)
handler_defs = re.findall(r'async def (_handle_\w+)\(self.*?\):\s*\n\s*"""', server_content, re.DOTALL)
all_handlers = re.findall(r'async def (_handle_\w+)\(', server_content)
doc_ratio = len(handler_defs) / max(len(all_handlers), 1)
record("DOC-Q4", "API handlers have docstrings", doc_ratio >= 0.3,
       f"{len(handler_defs)}/{len(all_handlers)} handlers documented ({doc_ratio:.0%})")

# Q5: Do docs use consistent structure (title, meta, breadcrumb)?
consistent_docs = 0
for df in doc_files:
    dc = read_file(df)
    has_title = bool(re.search(r'<title>', dc))
    has_meta = bool(re.search(r'<meta name="description"', dc))
    has_breadcrumb = bool(re.search(r'breadcrumb|docs-breadcrumb', dc))
    if has_title and has_meta and has_breadcrumb:
        consistent_docs += 1
record("DOC-Q5", "docs use consistent structure",
       consistent_docs >= len(doc_files) * 0.5 if doc_files else False,
       f"{consistent_docs}/{len(doc_files)} pages have title+meta+breadcrumb")

# Q6: Are there Mermaid diagrams for architecture?
diagram_dir = DATA / "diagrams"
diagrams = sorted(diagram_dir.glob("*.md")) if diagram_dir.is_dir() else []
mermaid_diagrams = []
for d in diagrams[:20]:
    dc = read_file(d)
    if re.search(r'```mermaid|graph |sequenceDiagram|stateDiagram|flowchart', dc):
        mermaid_diagrams.append(d.name)
record("DOC-Q6", "Mermaid architecture diagrams exist", len(mermaid_diagrams) >= 3,
       f"{len(mermaid_diagrams)} Mermaid diagrams found")

# Q7: CONTRIBUTING guide for AI agent developers?
contrib_path = BROWSER_ROOT / "CONTRIBUTING.md"
contrib_exists = file_exists(contrib_path)
contrib_content = read_file(contrib_path)
has_agent_section = bool(re.search(r'agent|MCP|AI|automation', contrib_content, re.IGNORECASE)) if contrib_content else False
record("DOC-Q7", "CONTRIBUTING.md mentions AI agents/MCP", contrib_exists and has_agent_section,
       f"exists={contrib_exists}, agent_section={has_agent_section}")
````

````python
# ============================================================
# Probe 7: Evidence API (Floors 43-49)
# Personas: Ethics Auditor, Model Evaluation, Legacy System,
#   Multi-Language Code, Safety Guardian, Dragon Rider, Mum Test Judge
# Questions probed:
#   Q1: Does src/evidence/__init__.py export seal_evidence and verify_chain?
#   Q2: Does the evidence module use SHA-256 hash chaining?
#   Q3: Is there an AuditChain class in src/audit/chain.py?
#   Q4: Does evidence_pipeline.py exist for batch evidence processing?
#   Q5: Can agents access evidence via /api/evidence/search endpoint?
#   Q6: Can agents verify evidence via /api/v1/evidence/verify endpoint?
#   Q7: Does evidence include timestamps and previous_hash for chain integrity?
# ============================================================

print("=" * 60)
print("PROBE 7: Evidence API (Floors 43-49)")
print("=" * 60)

ev_content = read_file(EVIDENCE_INIT)

# Q1: Does evidence module export seal_evidence and verify_chain?
has_seal = bool(re.search(r'def seal_evidence', ev_content))
has_verify = bool(re.search(r'def verify_chain', ev_content))
record("EVIDENCE-Q1", "evidence module exports seal_evidence + verify_chain",
       has_seal and has_verify,
       f"seal_evidence={has_seal}, verify_chain={has_verify}")

# Q2: Does evidence use SHA-256 hash chaining?
has_sha256 = bool(re.search(r'sha256|SHA-256|hashlib\.sha256', ev_content))
has_chain = bool(re.search(r'previous_hash|hash.*chain|genesis', ev_content))
record("EVIDENCE-Q2", "evidence uses SHA-256 hash chaining", has_sha256 and has_chain,
       f"sha256={has_sha256}, chain_linking={has_chain}")

# Q3: Is there an AuditChain class?
audit_content = read_file(AUDIT_CHAIN)
has_audit_chain = bool(re.search(r'class AuditChain', audit_content))
audit_methods = re.findall(r'def (\w+)', audit_content) if audit_content else []
record("EVIDENCE-Q3", "AuditChain class exists in audit/chain.py",
       has_audit_chain,
       f"class={has_audit_chain}, methods={audit_methods[:8]}")

# Q4: Does evidence_pipeline.py exist?
pipeline_exists = file_exists(EVIDENCE_PIPELINE)
pipeline_content = read_file(EVIDENCE_PIPELINE)
pipeline_lines = len(pipeline_content.splitlines()) if pipeline_content else 0
record("EVIDENCE-Q4", "evidence_pipeline.py exists for batch processing",
       pipeline_exists and pipeline_lines > 20,
       f"exists={pipeline_exists}, {pipeline_lines} lines")

# Q5: /api/evidence/search endpoint accessible to agents?
server_content = read_file(BROWSER_SERVER)
has_ev_search = bool(re.search(r'/api/evidence/search', server_content))
has_ev_search_handler = bool(re.search(r'_handle_evidence_search', server_content))
record("EVIDENCE-Q5", "/api/evidence/search endpoint exists",
       has_ev_search and has_ev_search_handler,
       f"route={has_ev_search}, handler={has_ev_search_handler}")

# Q6: /api/v1/evidence/verify endpoint?
has_ev_verify = bool(re.search(r'/api/v1/evidence/verify', server_content))
has_ev_verify_handler = bool(re.search(r'_handle_evidence_verify', server_content))
record("EVIDENCE-Q6", "/api/v1/evidence/verify endpoint exists",
       has_ev_verify and has_ev_verify_handler,
       f"route={has_ev_verify}, handler={has_ev_verify_handler}")

# Q7: Evidence entries include timestamps and previous_hash?
has_timestamp = bool(re.search(r'timestamp.*datetime|datetime.*timestamp|isoformat', ev_content))
has_prev_hash = bool(re.search(r'previous_hash', ev_content))
has_genesis = bool(re.search(r'GENESIS_HASH|genesis', ev_content))
record("EVIDENCE-Q7", "evidence includes timestamp + previous_hash",
       has_timestamp and has_prev_hash and has_genesis,
       f"timestamp={has_timestamp}, previous_hash={has_prev_hash}, genesis={has_genesis}")
````
