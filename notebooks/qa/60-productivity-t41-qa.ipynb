```python
# NORTHSTAR -- Tower 41 Productivity Suite QA for Solace Browser
# DNA: productivity(suite) = brief(morning) x calendar(sync) x focus(timer) x digest(weekly) x routine(automate) x insight(analyze) x time(save)
# Probes productivity recipes, schedule automation, workspace features
import os, sys, re, json, hashlib
from pathlib import Path
from datetime import datetime

NORTHSTAR = "productivity-t41-qa"
NOTEBOOK_PATH = "notebooks/qa/60-productivity-t41-qa.ipynb"
TOWER = 41
TOWER_NAME = "Tower of Productivity Suite"
MIN_SCORE = 70

BROWSER_ROOT = Path("/home/phuc/projects/solace-browser")
SRC = BROWSER_ROOT / "src"
WEB = BROWSER_ROOT / "web"
TESTS = BROWSER_ROOT / "tests"
DATA = BROWSER_ROOT / "data" / "default"

results = []

def record(probe_id, name, passed, detail=""):
    status = "PASS" if passed else "FAIL"
    results.append({"id": probe_id, "name": name, "status": status, "detail": detail})
    print(f"{status}: {name}" + (f" -- {detail}" if detail else ""))

def read_file(path):
    p = Path(path)
    return p.read_text(encoding='utf-8', errors='replace') if p.exists() else ""

print(f"Tower {TOWER}: {TOWER_NAME}")
assert BROWSER_ROOT.exists()
```

```python
# F1 -- Clara Richter: Morning Briefing
# Morning brief recipe must exist with structured steps

# Probe: Morning brief recipe exists
morning_recipe = read_file(DATA / "apps" / "morning-brief" / "recipe.json")
has_morning = len(morning_recipe) > 50
record("F1-P1", "Morning brief recipe exists", has_morning,
       f"{len(morning_recipe)} bytes")

# Probe: Morning brief budget exists
morning_budget = read_file(DATA / "apps" / "morning-brief" / "budget.json")
has_morning_budget = len(morning_budget) > 20
record("F1-P2", "Morning brief budget.json exists", has_morning_budget,
       f"{len(morning_budget)} bytes")

# Probe: Gmail inbox triage recipe (feeds morning brief)
gmail_recipe = read_file(DATA / "apps" / "gmail-inbox-triage" / "recipe.json")
has_gmail = len(gmail_recipe) > 50
record("F1-P3", "Gmail inbox triage recipe exists (feeds briefing)", has_gmail,
       f"{len(gmail_recipe)} bytes")
```

```python
# F2 -- James Oduya: Calendar Sync
# Calendar brief recipe + schedule system

# Probe: Calendar brief recipe exists
cal_recipe = read_file(DATA / "apps" / "calendar-brief" / "recipe.json")
has_cal = len(cal_recipe) > 50
record("F2-P1", "Calendar brief recipe exists", has_cal,
       f"{len(cal_recipe)} bytes")

# Probe: Schedule page exists
schedule_html = read_file(WEB / "schedule.html")
has_schedule = len(schedule_html) > 200
record("F2-P2", "Schedule page exists", has_schedule,
       f"{len(schedule_html)} bytes")

# Probe: Schedule calendar JS with time handling
sched_cal_js = read_file(WEB / "js" / "schedule-calendar.js")
has_time = bool(re.search(r'time|hour|minute|date|calendar', sched_cal_js, re.IGNORECASE))
record("F2-P3", "Schedule calendar JS handles time/date", has_time,
       f"{len(sched_cal_js)} bytes")

# F3 -- Sophie Andersen: Focus Sessions
# Focus timer recipe exists
focus_recipe = read_file(DATA / "apps" / "focus-timer" / "recipe.json")
has_focus = len(focus_recipe) > 50
record("F3-P1", "Focus timer recipe exists", has_focus,
       f"{len(focus_recipe)} bytes")

# Probe: Focus timer budget exists
focus_budget = read_file(DATA / "apps" / "focus-timer" / "budget.json")
has_focus_budget = len(focus_budget) > 20
record("F3-P2", "Focus timer budget.json exists", has_focus_budget,
       f"{len(focus_budget)} bytes")
```

```python
# F4 -- Patrick Leung: Weekly Digest

# Probe: Weekly digest recipe exists
digest_recipe = read_file(DATA / "apps" / "weekly-digest" / "recipe.json")
has_digest = len(digest_recipe) > 50
record("F4-P1", "Weekly digest recipe exists", has_digest,
       f"{len(digest_recipe)} bytes")

# Probe: Weekly digest budget exists
digest_budget = read_file(DATA / "apps" / "weekly-digest" / "budget.json")
has_digest_budget = len(digest_budget) > 20
record("F4-P2", "Weekly digest budget.json exists", has_digest_budget,
       f"{len(digest_budget)} bytes")

# F5 -- Maria Gonzalez: Routine Automation
# Recipe compiler and executor for chained workflows

# Probe: Recipe compiler supports multi-step workflows
compiler = read_file(SRC / "recipes" / "recipe_compiler.py")
has_compiler = bool(re.search(r'step|chain|sequence|pipeline|compile', compiler, re.IGNORECASE))
record("F5-P1", "Recipe compiler supports multi-step workflows", has_compiler,
       f"{len(compiler)} bytes")

# Probe: Recipe executor exists
executor = read_file(SRC / "recipes" / "recipe_executor.py")
has_executor = bool(re.search(r'execut|run|process|step', executor, re.IGNORECASE))
record("F5-P2", "Recipe executor module exists", has_executor,
       f"{len(executor)} bytes")

# Probe: Slack triage recipe (routine automation example)
slack_recipe = read_file(DATA / "apps" / "slack-triage" / "recipe.json")
has_slack = len(slack_recipe) > 50
record("F5-P3", "Slack triage recipe exists (routine example)", has_slack,
       f"{len(slack_recipe)} bytes")
```

```python
# NEGATIVE SPACE (P16) -- Productivity infrastructure

# Probe: Inbox/outbox system for async task handling
inbox = read_file(SRC / "inbox_outbox.py")
has_inbox = bool(re.search(r'inbox|outbox|queue|pending', inbox, re.IGNORECASE))
record("NS-P1", "Inbox/outbox system for async task handling", has_inbox,
       f"{len(inbox)} bytes")

# Probe: Google Drive saver recipe (productivity tool)
drive_recipe = read_file(DATA / "apps" / "google-drive-saver" / "recipe.json")
has_drive = len(drive_recipe) > 50
record("NS-P2", "Google Drive saver recipe exists", has_drive,
       f"{len(drive_recipe)} bytes")

# Probe: Schedule approvals for human-in-the-loop
approvals_js = read_file(WEB / "js" / "schedule-approvals.js")
has_approvals = bool(re.search(r'approv|confirm|review|gate', approvals_js, re.IGNORECASE))
record("NS-P3", "Schedule approvals JS module exists", has_approvals,
       f"{len(approvals_js)} bytes")

# Probe: Approvals test exists
approvals_test = TESTS / "test_approvals.py"
has_approv_test = approvals_test.exists() and approvals_test.stat().st_size > 100
record("NS-P4", "Approvals test exists", has_approv_test)
```

```python
# EVIDENCE SUMMARY -- Tower 41 Productivity Suite QA
passed = len([r for r in results if r["status"] == "PASS"])
failed = len([r for r in results if r["status"] == "FAIL"])
total = len(results)
score = round(passed / total * 100, 1) if total > 0 else 0
finding_rate = round(failed / total * 100, 1) if total > 0 else 0
grade = "A+" if score >= 95 else "A" if score >= 90 else "B+" if score >= 85 else "B" if score >= 80 else "C" if score >= 70 else "D" if score >= 60 else "F"

summary = {
    "surface": NORTHSTAR, "tower": TOWER, "tower_name": TOWER_NAME,
    "timestamp": datetime.now().isoformat(),
    "total_probes": total, "passed": passed, "failed": failed,
    "score": score, "grade": grade, "finding_rate": finding_rate,
    "converged": finding_rate < 5,
    "findings": [r for r in results if r["status"] == "FAIL"],
    "evidence_hash": hashlib.sha256(json.dumps(results, sort_keys=True).encode()).hexdigest()[:16]
}

print("=" * 60)
print(f"TOWER {TOWER}: {TOWER_NAME}")
print("=" * 60)
print(f"Probes: {total} | Passed: {passed} | Failed: {failed}")
print(f"Score: {score}% | Grade: {grade} | Finding rate: {finding_rate}%")
print(f"Converged: {'YES' if summary['converged'] else 'NO'}")
print(f"Evidence hash: {summary['evidence_hash']}")
if summary["findings"]:
    print("\nFINDINGS:")
    for f in summary["findings"]:
        print(f"  [{f['id']}] {f['name']}: {f.get('detail', '')}")
print()
print(json.dumps(summary, indent=2))
```
