```python
# NORTHSTAR -- Tower 30 YinYang Assistant + Onboarding QA for Solace Browser
# DNA: yinyang(assistant) = greet(welcome) x tutorial(5steps) x chat(context) x suggest(actions) x track(credits) x delight(surprise) x persist(memory)
# Probes the YinYang assistant, onboarding tutorial, and first-time user experience
import os, sys, re, json, hashlib
from pathlib import Path
from datetime import datetime

NORTHSTAR = "yinyang-onboarding-t30-qa"
NOTEBOOK_PATH = "notebooks/qa/17-yinyang-onboarding-t30-qa.ipynb"
TOWER = 30
TOWER_NAME = "Tower of YinYang Assistant + Onboarding"
MIN_SCORE = 70

BROWSER_ROOT = Path("/home/phuc/projects/solace-browser")
SRC = BROWSER_ROOT / "src"
WEB = BROWSER_ROOT / "web"
TESTS = BROWSER_ROOT / "tests"
YINYANG = SRC / "yinyang"

results = []

def record(probe_id, name, passed, detail=""):
    status = "PASS" if passed else "FAIL"
    results.append({"id": probe_id, "name": name, "status": status, "detail": detail})
    print(f"{status}: {name}" + (f" -- {detail}" if detail else ""))

def read_file(path):
    p = Path(path)
    return p.read_text(encoding='utf-8', errors='replace') if p.exists() else ""

print(f"Tower {TOWER}: {TOWER_NAME}")
assert BROWSER_ROOT.exists()
```

```python
# F1 -- Grace Okafor: First-Time Welcome
# Q: Does YinYang appear with friendly welcome within 3 seconds?
# Q: Does welcome explain Solace in one sentence without jargon?

# Probe: YinYang personality module exists with greeting logic
personality_code = read_file(YINYANG / "personality.py")
has_greeting = bool(re.search(r'greet|welcome|hello|first.*time|onboard', personality_code, re.IGNORECASE))
record("F1-P1", "YinYang personality has greeting/welcome logic", has_greeting,
       f"{len(personality_code)} bytes")

# Probe: delight engine exists for warm interactions
delight_code = read_file(YINYANG / "delight_engine.py")
has_delight = len(delight_code) > 100
record("F1-P2", "Delight engine module exists", has_delight,
       f"{len(delight_code)} bytes")

# Probe: bottom rail (persistent assistant UI)
bottom_rail = read_file(YINYANG / "bottom_rail.py")
has_rail = bool(re.search(r'show|hide|toggle|render|display', bottom_rail, re.IGNORECASE))
record("F1-P3", "Bottom rail UI module exists", has_rail)

# Probe: state bridge tracks user state (first time vs returning)
state_bridge = read_file(YINYANG / "state_bridge.py")
has_state = bool(re.search(r'state|first.*time|return|new.*user|seen', state_bridge, re.IGNORECASE))
record("F1-P4", "State bridge tracks first-time vs returning user", has_state,
       f"{len(state_bridge)} bytes")
```

```python
# F2 -- Ben Nakamura: Tutorial Walkthrough
# Q: Does tutorial walk through 5 steps?
# Q: Can user skip/go back in tutorial?

# Probe: tutorial JS file exists with step logic
tutorial_v2 = read_file(WEB / "js" / "yinyang-tutorial-v2.js")
tutorial_v1 = read_file(WEB / "js" / "yinyang-tutorial.js")
tutorial_code = tutorial_v2 if len(tutorial_v2) > len(tutorial_v1) else tutorial_v1
has_steps = bool(re.search(r'step|next|prev|back|skip|progress', tutorial_code, re.IGNORECASE))
record("F2-P1", "Tutorial JS has step navigation", has_steps,
       f"v2={len(tutorial_v2)}B, v1={len(tutorial_v1)}B")

# Probe: tutorial has at least 3 steps defined
step_defs = re.findall(r'step[_\s]*\d|STEPS\[|steps\.push|step.*title', tutorial_code, re.IGNORECASE)
record("F2-P2", "Tutorial defines multiple steps", len(step_defs) >= 3,
       f"{len(step_defs)} step references")

# Probe: onboarding JS exists
onboarding_code = read_file(WEB / "js" / "onboarding.js")
has_onboarding = len(onboarding_code) > 100
record("F2-P3", "Onboarding JS module exists", has_onboarding,
       f"{len(onboarding_code)} bytes")

# Probe: guide page exists for getting started
guide_html = read_file(WEB / "guide.html")
has_guide = len(guide_html) > 200
record("F2-P4", "Getting started guide page exists", has_guide,
       f"{len(guide_html)} bytes")
```

```python
# F3 -- Isabelle Fontaine: Context-Aware Chat
# Q: Can assistant respond with page-aware context?
# Q: Does assistant know user's tier?

# Probe: websocket bridge for real-time chat
ws_bridge = read_file(YINYANG / "ws_bridge.py")
has_ws = bool(re.search(r'websocket|ws|connect|message|send|receive', ws_bridge, re.IGNORECASE))
record("F3-P1", "WebSocket bridge for real-time chat exists", has_ws,
       f"{len(ws_bridge)} bytes")

# Probe: top rail provides contextual suggestions
top_rail = read_file(YINYANG / "top_rail.py")
has_context = bool(re.search(r'context|suggest|page|current|recommend', top_rail, re.IGNORECASE))
record("F3-P2", "Top rail provides contextual suggestions", has_context,
       f"{len(top_rail)} bytes")

# Probe: highlighter for UI element highlighting
highlighter = read_file(YINYANG / "highlighter.py")
has_highlight = bool(re.search(r'highlight|spotlight|focus|element|pointer', highlighter, re.IGNORECASE))
record("F3-P3", "Highlighter module for UI hints exists", has_highlight,
       f"{len(highlighter)} bytes")

# Probe: support bridge for help requests
support_bridge = read_file(YINYANG / "support_bridge.py")
has_support = bool(re.search(r'support|help|ticket|assist', support_bridge, re.IGNORECASE))
record("F3-P4", "Support bridge for help requests exists", has_support)
```

```python
# F4 -- Omar Hassan: Smart Suggestions
# Q: Does assistant suggest relevant actions based on page?

# Probe: alert queue for push suggestions
alert_queue = read_file(YINYANG / "alert_queue.py")
has_alerts = bool(re.search(r'alert|queue|push|notify|suggest', alert_queue, re.IGNORECASE))
record("F4-P1", "Alert queue for push suggestions exists", has_alerts,
       f"{len(alert_queue)} bytes")

# Probe: push alerts module
push_alerts = read_file(YINYANG / "push_alerts.py")
has_push = len(push_alerts) > 100
record("F4-P2", "Push alerts module exists", has_push,
       f"{len(push_alerts)} bytes")

# Probe: yinyang-rail.js for frontend rail rendering
rail_js = read_file(WEB / "js" / "yinyang-rail.js")
has_rail_js = bool(re.search(r'rail|assistant|chat|message', rail_js, re.IGNORECASE))
record("F4-P3", "YinYang rail JS for frontend rendering exists", has_rail_js,
       f"{len(rail_js)} bytes")
```

```python
# F5 -- Credits Tracking + F6 -- Delight/Surprise
# Q: Can user see credit usage? Does assistant delight?

# Probe: yinyang-delight.js exists for delight animations
delight_js = read_file(WEB / "js" / "yinyang-delight.js")
has_delight_js = bool(re.search(r'delight|celebration|confetti|animate|surprise', delight_js, re.IGNORECASE))
record("F5-P1", "Delight JS for celebrations/animations exists", has_delight_js,
       f"{len(delight_js)} bytes")

# Probe: settings page shows user preferences for assistant
settings_html = read_file(WEB / "settings.html")
has_assistant_settings = bool(re.search(r'yinyang|assistant|notification|alert', settings_html, re.IGNORECASE))
record("F5-P2", "Settings page has assistant preferences", has_assistant_settings)

# Probe: yinyang tests exist
yinyang_tests = list(TESTS.glob("test_yinyang_*.py"))
record("F5-P3", "YinYang test suite exists", len(yinyang_tests) >= 2,
       f"{len(yinyang_tests)} yinyang test files")
```

```python
# F7 -- Persistence (Memory)
# Q: Does assistant remember across sessions?
# NEGATIVE SPACE (P16) -- What's missing?

# Probe: YinYang __init__.py exports key modules
yy_init = read_file(YINYANG / "__init__.py")
has_exports = len(yy_init) > 10
record("F7-P1", "YinYang package properly initialized", has_exports,
       f"{len(yy_init)} bytes")

# Probe: OAuth3 consent confirmation for assistant actions
oauth3_confirm = read_file(WEB / "js" / "yinyang-oauth3-confirm.js")
has_consent_js = bool(re.search(r'confirm|consent|approve|oauth3|scope', oauth3_confirm, re.IGNORECASE))
record("F7-P2", "YinYang OAuth3 consent confirmation JS exists", has_consent_js,
       f"{len(oauth3_confirm)} bytes")

# NEGATIVE SPACE: No bare excepts in YinYang module
yy_files = list(YINYANG.glob("*.py")) if YINYANG.exists() else []
bare_excepts = 0
for yf in yy_files:
    content = yf.read_text(encoding='utf-8', errors='replace')
    bare_excepts += len(re.findall(r'^\s*except\s*:', content, re.MULTILINE))
    bare_excepts += len(re.findall(r'^\s*except\s+Exception\s*:', content, re.MULTILINE))
record("NS-P1", "No bare excepts in YinYang (Fallback Ban)",
       bare_excepts == 0, f"{bare_excepts} bare excepts across {len(yy_files)} files")

# NEGATIVE SPACE: start.html has onboarding triggers
start_html = read_file(WEB / "start.html")
has_start_trigger = bool(re.search(r'tutorial|onboard|wizard|setup|welcome', start_html, re.IGNORECASE))
record("NS-P2", "start.html triggers onboarding", has_start_trigger)
```

```python
# EVIDENCE SUMMARY -- Tower 30 YinYang + Onboarding QA
passed = len([r for r in results if r["status"] == "PASS"])
failed = len([r for r in results if r["status"] == "FAIL"])
total = len(results)
score = round(passed / total * 100, 1) if total > 0 else 0
finding_rate = round(failed / total * 100, 1) if total > 0 else 0
grade = "A+" if score >= 95 else "A" if score >= 90 else "B+" if score >= 85 else "B" if score >= 80 else "C" if score >= 70 else "D" if score >= 60 else "F"

summary = {
    "surface": NORTHSTAR, "tower": TOWER, "tower_name": TOWER_NAME,
    "timestamp": datetime.now().isoformat(),
    "total_probes": total, "passed": passed, "failed": failed,
    "score": score, "grade": grade, "finding_rate": finding_rate,
    "converged": finding_rate < 5,
    "findings": [r for r in results if r["status"] == "FAIL"],
    "evidence_hash": hashlib.sha256(json.dumps(results, sort_keys=True).encode()).hexdigest()[:16]
}

print("=" * 60)
print(f"TOWER {TOWER}: {TOWER_NAME}")
print("=" * 60)
print(f"Probes: {total} | Passed: {passed} | Failed: {failed}")
print(f"Score: {score}% | Grade: {grade} | Finding rate: {finding_rate}%")
print(f"Converged: {'YES' if summary['converged'] else 'NO'}")
print(f"Evidence hash: {summary['evidence_hash']}")
if summary["findings"]:
    print("\nFINDINGS:")
    for f in summary["findings"]:
        print(f"  [{f['id']}] {f['name']}: {f.get('detail', '')}")
print()
print(json.dumps(summary, indent=2))
```
