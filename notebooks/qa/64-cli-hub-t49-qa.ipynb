```python
# NORTHSTAR -- Tower 49 CLI Hub Integration QA for Solace Browser
# DNA: cli_hub(integration) = browser_control(spawn_monitor) x cloud_deploy(push_sync) x skill_store(private_sealed) x recipe_author(create_test_seal) x stillwater_ext(oss_private_boundary) x remote_target(dispatch_security) x evidence(audit_all)
# Probes CLI integration points, API endpoints, token exchange, recipe format, evidence format
import os, sys, re, json, hashlib
from pathlib import Path
from datetime import datetime

NORTHSTAR = "cli-hub-t49-qa"
NOTEBOOK_PATH = "notebooks/qa/64-cli-hub-t49-qa.ipynb"
TOWER = 49
TOWER_NAME = "Tower of CLI Hub Integration"
MIN_SCORE = 70

BROWSER_ROOT = Path("/home/phuc/projects/solace-browser")
SRC = BROWSER_ROOT / "src"
WEB = BROWSER_ROOT / "web"
TESTS = BROWSER_ROOT / "tests"
DATA = BROWSER_ROOT / "data" / "default"

results = []

def record(probe_id, name, passed, detail=""):
    status = "PASS" if passed else "FAIL"
    results.append({"id": probe_id, "name": name, "status": status, "detail": detail})
    print(f"{status}: {name}" + (f" -- {detail}" if detail else ""))

def read_file(path):
    p = Path(path)
    return p.read_text(encoding='utf-8', errors='replace') if p.exists() else ""

print(f"Tower {TOWER}: {TOWER_NAME}")
assert BROWSER_ROOT.exists()
```

```python
# F1 -- Marcus Chen: CLI Browser Control
# CLI module, browser context, machine API for headless control

# Probe: CLI module exists with command handling
cli_code = read_file(SRC / "cli" / "cli.py")
has_cli = bool(re.search(r'command|argument|parse|run|cli', cli_code, re.IGNORECASE))
record("F1-P1", "CLI module exists with command handling", has_cli,
       f"{len(cli_code)} bytes")

# Probe: CLI test exists
cli_test = TESTS / "test_cli.py"
has_cli_test = cli_test.exists() and cli_test.stat().st_size > 100
record("F1-P2", "CLI test exists", has_cli_test)

# Probe: Browser context module for session management
browser_ctx = read_file(SRC / "browser" / "context.py")
has_ctx = bool(re.search(r'context|session|browser|page|navigate', browser_ctx, re.IGNORECASE))
record("F1-P3", "Browser context module exists", has_ctx,
       f"{len(browser_ctx)} bytes")

# Probe: Machine API for headless browser control
machine_api = read_file(SRC / "machine" / "api.py")
has_machine = bool(re.search(r'api|endpoint|route|machine|headless', machine_api, re.IGNORECASE))
record("F1-P4", "Machine API for headless control exists", has_machine,
       f"{len(machine_api)} bytes")

# Probe: Machine tunnel module for remote connectivity
tunnel_code = read_file(SRC / "machine" / "tunnel.py")
has_tunnel = bool(re.search(r'tunnel|connect|remote|port|proxy', tunnel_code, re.IGNORECASE))
record("F1-P5", "Machine tunnel module exists", has_tunnel,
       f"{len(tunnel_code)} bytes")
```

```python
# F2 -- Elena Rodriguez: CLI Cloud Deploy
# Dockerfile, CI workflows, cloud runner

# Probe: Dockerfile for cloud deployment
dockerfile = read_file(BROWSER_ROOT / "Dockerfile")
has_docker = bool(re.search(r'FROM|EXPOSE|CMD|ENTRYPOINT', dockerfile))
record("F2-P1", "Dockerfile exists for deployment", has_docker,
       f"{len(dockerfile)} bytes")

# Probe: CI workflow exists
ci_yml = read_file(BROWSER_ROOT / ".github" / "workflows" / "ci.yml")
has_ci = bool(re.search(r'test|build|deploy|push', ci_yml, re.IGNORECASE))
record("F2-P2", "CI workflow exists", has_ci,
       f"{len(ci_yml)} bytes")

# Probe: Cloud runner cloud_app.py exists
cloud_app = read_file(SRC / "cloud_runner" / "cloud_app.py")
has_cloud = bool(re.search(r'execute|health|app|route', cloud_app, re.IGNORECASE))
record("F2-P3", "Cloud runner app with execute endpoint exists", has_cloud,
       f"{len(cloud_app)} bytes")

# Probe: Cloud test support module exists
cloud_support = read_file(TESTS / "cloud_test_support.py")
has_support = len(cloud_support) > 50
record("F2-P4", "Cloud test support module exists", has_support,
       f"{len(cloud_support)} bytes")
```

```python
# F3 -- David Park: Private Skill Store
# Store client, credential vault for auth

# Probe: Store client module exists
store_client = read_file(SRC / "store_client" / "store_client.py")
has_store = bool(re.search(r'store|publish|fetch|version|skill', store_client, re.IGNORECASE))
record("F3-P1", "Store client module exists", has_store,
       f"{len(store_client)} bytes")

# Probe: Store client test exists
store_test = TESTS / "test_store_client.py"
has_store_test = store_test.exists() and store_test.stat().st_size > 100
record("F3-P2", "Store client test exists", has_store_test)

# Probe: Credential vault for secure auth storage
cred_vault = read_file(SRC / "cli" / "credential_vault.py")
has_cred = bool(re.search(r'vault|credential|token|secret|key', cred_vault, re.IGNORECASE))
record("F3-P3", "Credential vault module exists", has_cred,
       f"{len(cred_vault)} bytes")

# Probe: Credential vault test exists
cred_test = TESTS / "test_credential_vault.py"
has_cred_test = cred_test.exists() and cred_test.stat().st_size > 100
record("F3-P4", "Credential vault test exists", has_cred_test)
```

```python
# F4 -- Amara Osei: Recipe Authoring
# Recipe parser, compiler, executor, seal

# Probe: Recipe parser exists
parser = read_file(SRC / "recipes" / "recipe_parser.py")
has_parser = bool(re.search(r'parse|validate|load|schema', parser, re.IGNORECASE))
record("F4-P1", "Recipe parser module exists", has_parser,
       f"{len(parser)} bytes")

# Probe: Recipe compiler exists
compiler = read_file(SRC / "recipes" / "recipe_compiler.py")
has_compiler = bool(re.search(r'compile|transform|step|ir|intermediate', compiler, re.IGNORECASE))
record("F4-P2", "Recipe compiler module exists", has_compiler,
       f"{len(compiler)} bytes")

# Probe: Recipe executor exists
executor = read_file(SRC / "recipes" / "recipe_executor.py")
has_executor = bool(re.search(r'execut|run|step|action', executor, re.IGNORECASE))
record("F4-P3", "Recipe executor module exists", has_executor,
       f"{len(executor)} bytes")

# Probe: Recipe JSON files use consistent format
recipe_files = list(DATA.rglob("recipe.json"))
valid_recipes = 0
for rf in recipe_files:
    content = rf.read_text(encoding='utf-8', errors='replace')
    try:
        data = json.loads(content)
        if "name" in data or "steps" in data or "recipe_name" in data:
            valid_recipes += 1
    except json.JSONDecodeError:
        pass
record("F4-P4", "Recipe JSON files with valid structure (>=10)",
       valid_recipes >= 10, f"{valid_recipes}/{len(recipe_files)} valid")
```

```python
# EVIDENCE SUMMARY -- Tower 49 CLI Hub Integration QA
passed = len([r for r in results if r["status"] == "PASS"])
failed = len([r for r in results if r["status"] == "FAIL"])
total = len(results)
score = round(passed / total * 100, 1) if total > 0 else 0
finding_rate = round(failed / total * 100, 1) if total > 0 else 0
grade = "A+" if score >= 95 else "A" if score >= 90 else "B+" if score >= 85 else "B" if score >= 80 else "C" if score >= 70 else "D" if score >= 60 else "F"

summary = {
    "surface": NORTHSTAR, "tower": TOWER, "tower_name": TOWER_NAME,
    "timestamp": datetime.now().isoformat(),
    "total_probes": total, "passed": passed, "failed": failed,
    "score": score, "grade": grade, "finding_rate": finding_rate,
    "converged": finding_rate < 5,
    "findings": [r for r in results if r["status"] == "FAIL"],
    "evidence_hash": hashlib.sha256(json.dumps(results, sort_keys=True).encode()).hexdigest()[:16]
}

print("=" * 60)
print(f"TOWER {TOWER}: {TOWER_NAME}")
print("=" * 60)
print(f"Probes: {total} | Passed: {passed} | Failed: {failed}")
print(f"Score: {score}% | Grade: {grade} | Finding rate: {finding_rate}%")
print(f"Converged: {'YES' if summary['converged'] else 'NO'}")
print(f"Evidence hash: {summary['evidence_hash']}")
if summary["findings"]:
    print("\nFINDINGS:")
    for f in summary["findings"]:
        print(f"  [{f['id']}] {f['name']}: {f.get('detail', '')}")
print()
print(json.dumps(summary, indent=2))
```
