```python
# NORTHSTAR -- Tower 44 YinYang AI Assistant QA for Solace Browser
# DNA: yinyang(assistant) = chat(persistent) x suggest(contextual) x help(inline) x action(quick) x learn(user) x delight(warm) x compete(copilot)
# Probes chat persistence, contextual suggestions, inline help, quick actions, learning, delight
import os, sys, re, json, hashlib
from pathlib import Path
from datetime import datetime

NORTHSTAR = "yinyang-ai-t44-qa"
NOTEBOOK_PATH = "notebooks/qa/22-yinyang-ai-t44-qa.ipynb"
TOWER = 44
TOWER_NAME = "Tower of YinYang AI Assistant"
MIN_SCORE = 70

BROWSER_ROOT = Path("/home/phuc/projects/solace-browser")
SRC = BROWSER_ROOT / "src"
WEB = BROWSER_ROOT / "web"
TESTS = BROWSER_ROOT / "tests"
YINYANG = SRC / "yinyang"

results = []

def record(probe_id, name, passed, detail=""):
    status = "PASS" if passed else "FAIL"
    results.append({"id": probe_id, "name": name, "status": status, "detail": detail})
    print(f"{status}: {name}" + (f" -- {detail}" if detail else ""))

def read_file(path):
    p = Path(path)
    return p.read_text(encoding='utf-8', errors='replace') if p.exists() else ""

print(f"Tower {TOWER}: {TOWER_NAME}")
assert BROWSER_ROOT.exists()
```

```python
# F1 -- Maya Rodriguez: First-Time Chat (Beginner)
# Q: Is there a visible chat bar? Does it respond within 3s?

# Probe: YinYang rail JS renders chat interface
rail_js = read_file(WEB / "js" / "yinyang-rail.js")
has_chat_ui = bool(re.search(r'chat|input|message|rail|send|type', rail_js, re.IGNORECASE))
record("F1-P1", "YinYang rail JS renders chat interface", has_chat_ui,
       f"{len(rail_js)} bytes")

# Probe: chat is visible on main pages (check index.html for yinyang)
index_html = read_file(WEB / "index.html")
has_yinyang_ref = bool(re.search(r'yinyang|yin.?yang|assistant|rail', index_html, re.IGNORECASE))
record("F1-P2", "Index page references YinYang assistant", has_yinyang_ref)

# Probe: personality module provides friendly responses
personality = read_file(YINYANG / "personality.py")
has_friendly = bool(re.search(r'greet|friend|warm|help|welcome|tone', personality, re.IGNORECASE))
record("F1-P3", "Personality module has warm/friendly tone", has_friendly,
       f"{len(personality)} bytes")

# Probe: suggested questions/prompts for new users
has_suggestions = bool(re.search(r'suggest|prompt|example|starter|try.*ask', rail_js, re.IGNORECASE))
record("F1-P4", "Chat has suggested prompts for new users", has_suggestions)
```

```python
# F2 -- Jordan Blake: Contextual Suggestions (Power User)
# Q: Does assistant change suggestions based on current page?

# Probe: state bridge tracks current page/context
state_bridge = read_file(YINYANG / "state_bridge.py")
has_page_context = bool(re.search(r'page|context|current|route|url|location', state_bridge, re.IGNORECASE))
record("F2-P1", "State bridge tracks current page context", has_page_context,
       f"{len(state_bridge)} bytes")

# Probe: top rail provides page-specific suggestions
top_rail = read_file(YINYANG / "top_rail.py")
has_suggestions = bool(re.search(r'suggest|recommend|action|context|page', top_rail, re.IGNORECASE))
record("F2-P2", "Top rail provides contextual suggestions", has_suggestions,
       f"{len(top_rail)} bytes")

# Probe: delight engine for celebrations/rewards
delight = read_file(YINYANG / "delight_engine.py")
has_delight = bool(re.search(r'delight|celebrate|reward|achieve|milestone', delight, re.IGNORECASE))
record("F2-P3", "Delight engine for celebrations exists", has_delight,
       f"{len(delight)} bytes")
```

```python
# F3 -- Inline Help + F4 -- Quick Actions
# Q: Can I get help on the element I'm looking at?
# Q: Can the assistant execute actions (run recipe, open page)?

# Probe: highlighter for inline element help
highlighter = read_file(YINYANG / "highlighter.py")
has_highlight = bool(re.search(r'highlight|focus|element|tooltip|overlay', highlighter, re.IGNORECASE))
record("F3-P1", "Highlighter for inline element help exists", has_highlight,
       f"{len(highlighter)} bytes")

# Probe: WebSocket bridge for real-time communication
ws_bridge = read_file(YINYANG / "ws_bridge.py")
has_ws = bool(re.search(r'websocket|ws|connect|send|receive|message', ws_bridge, re.IGNORECASE))
record("F3-P2", "WebSocket bridge for real-time assistant communication", has_ws,
       f"{len(ws_bridge)} bytes")

# Probe: companion apps module for executing actions
companion_apps = read_file(SRC / "companion" / "apps.py")
has_actions = bool(re.search(r'action|execute|run|launch|install|trigger', companion_apps, re.IGNORECASE))
record("F4-P1", "Companion apps module supports action execution", has_actions,
       f"{len(companion_apps)} bytes")

# Probe: companion scopes limit assistant's abilities
companion_scopes = read_file(SRC / "companion" / "scopes.py")
has_scopes = bool(re.search(r'scope|permission|allow|deny|limit', companion_scopes, re.IGNORECASE))
record("F4-P2", "Companion scopes limit assistant actions", has_scopes,
       f"{len(companion_scopes)} bytes")
```

```python
# F5 -- Learning + F6 -- Delight/Warmth
# Q: Does the assistant learn from user behavior?
# Q: Is the assistant warm and non-intrusive?

# Probe: history module tracks past interactions
history = read_file(SRC / "history.py")
has_history = bool(re.search(r'history|past|previous|record|session', history, re.IGNORECASE))
record("F5-P1", "History module tracks past interactions", has_history,
       f"{len(history)} bytes")

# Probe: settings manager stores user preferences
settings = read_file(SRC / "settings_manager.py")
has_prefs = bool(re.search(r'preference|setting|config|save|load|theme|font', settings, re.IGNORECASE))
record("F5-P2", "Settings manager stores user preferences", has_prefs,
       f"{len(settings)} bytes")

# Probe: delight JS provides warm frontend interactions
delight_js = read_file(WEB / "js" / "yinyang-delight.js")
has_delight_ui = bool(re.search(r'delight|animate|celebrate|confetti|sparkle|warm', delight_js, re.IGNORECASE))
record("F6-P1", "Delight JS provides warm frontend interactions", has_delight_ui,
       f"{len(delight_js)} bytes")

# Probe: yinyang branding test
branding_test = TESTS / "test_yinyang_branding.py"
has_brand_test = branding_test.exists() and branding_test.stat().st_size > 100
record("F6-P2", "YinYang branding test exists", has_brand_test)
```

```python
# F7 -- Compete with Copilot
# NEGATIVE SPACE (P16)

# Probe: YinYang integration test covers full flow
integration_test = TESTS / "test_yinyang_integration.py"
has_int_test = integration_test.exists() and integration_test.stat().st_size > 100
record("F7-P1", "YinYang integration test exists", has_int_test)

# Probe: YinYang rails test
rails_test = TESTS / "test_yinyang_rails.py"
has_rails_test = rails_test.exists() and rails_test.stat().st_size > 100
record("F7-P2", "YinYang rails test exists", has_rails_test)

# Probe: YinYang delight test
delight_test = TESTS / "test_yinyang_delight.py"
has_del_test = delight_test.exists() and delight_test.stat().st_size > 100
record("F7-P3", "YinYang delight test exists", has_del_test)

# NEGATIVE SPACE: no bare excepts in YinYang
yy_files = list(YINYANG.glob("*.py")) if YINYANG.exists() else []
bare_excepts = 0
for yf in yy_files:
    content = yf.read_text(encoding='utf-8', errors='replace')
    bare_excepts += len(re.findall(r'^\s*except\s*:', content, re.MULTILINE))
    bare_excepts += len(re.findall(r'^\s*except\s+Exception\s*:', content, re.MULTILINE))
record("NS-P1", "No bare excepts in YinYang (Fallback Ban)",
       bare_excepts == 0, f"{bare_excepts} bare excepts across {len(yy_files)} files")

# NEGATIVE SPACE: competitive features module
competitive = read_file(SRC / "competitive_features.py")
has_competitive = len(competitive) > 100
record("NS-P2", "Competitive features module exists", has_competitive,
       f"{len(competitive)} bytes")
```

```python
# EVIDENCE SUMMARY -- Tower 44 YinYang AI Assistant QA
passed = len([r for r in results if r["status"] == "PASS"])
failed = len([r for r in results if r["status"] == "FAIL"])
total = len(results)
score = round(passed / total * 100, 1) if total > 0 else 0
finding_rate = round(failed / total * 100, 1) if total > 0 else 0
grade = "A+" if score >= 95 else "A" if score >= 90 else "B+" if score >= 85 else "B" if score >= 80 else "C" if score >= 70 else "D" if score >= 60 else "F"

summary = {
    "surface": NORTHSTAR, "tower": TOWER, "tower_name": TOWER_NAME,
    "timestamp": datetime.now().isoformat(),
    "total_probes": total, "passed": passed, "failed": failed,
    "score": score, "grade": grade, "finding_rate": finding_rate,
    "converged": finding_rate < 5,
    "findings": [r for r in results if r["status"] == "FAIL"],
    "evidence_hash": hashlib.sha256(json.dumps(results, sort_keys=True).encode()).hexdigest()[:16]
}

print("=" * 60)
print(f"TOWER {TOWER}: {TOWER_NAME}")
print("=" * 60)
print(f"Probes: {total} | Passed: {passed} | Failed: {failed}")
print(f"Score: {score}% | Grade: {grade} | Finding rate: {finding_rate}%")
print(f"Converged: {'YES' if summary['converged'] else 'NO'}")
print(f"Evidence hash: {summary['evidence_hash']}")
if summary["findings"]:
    print("\nFINDINGS:")
    for f in summary["findings"]:
        print(f"  [{f['id']}] {f['name']}: {f.get('detail', '')}")
print()
print(json.dumps(summary, indent=2))
```
