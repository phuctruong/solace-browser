```python
%%bash\nset -euo pipefail\ncd /home/phuc/projects/solace-browser\necho '=== BEFORE: dist release artifacts ==='\nls -l dist/solace-browser-linux-x86_64 dist/solace-browser-linux-x86_64.sha256 2>/dev/null || true\necho '=== BEFORE: latest release-cycle dirs ==='\nls -1 scratch/release-cycle | tail -n 5\n
```

```python
%%bash\nset -euo pipefail\ncd /home/phuc/projects/solace-browser\nTARGET_OS=linux UPLOAD_ENABLED=0 DOWNLOAD_ENABLED=0 RUN_SMOKE=0 src/scripts/release_browser_cycle.sh\n
```

```python
%%bash\nset -euo pipefail\ncd /home/phuc/projects/solace-browser\nLATEST=$(ls -1 scratch/release-cycle | tail -n 1)\necho "=== AFTER: latest run = ${LATEST} ==="\nls -la "scratch/release-cycle/${LATEST}"\necho '=== metrics.json ==='\ncat "scratch/release-cycle/${LATEST}/metrics.json"\n
```
